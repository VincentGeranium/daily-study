## 1.1 totalDistance가 0 이면서 Kills가 1 이상 = 비정상적 데이터

#### totalDistance(rideDistance + walkDistance + swimDistance)를 새로운 컬럼으로 만든다.
#### killsWithoutMoving은 totalDistace가 0 이면서 kills가 1 이상인 것.
#### killsWithoutMoving이 True 이면 hack으로 간주한다

````python
train['totalDistance'] = train['rideDistance'] + train['walkDistance'] + train['swimDistance']
train['killsWithoutMoving'] = ((train['kills'] > 0) & (train['totalDistance'] == 0))
````

## 1.5 longestKill이 비정상적으로 높으면서 kills도 비정상적으로 높은 값 = Aim Hack

## Aimbot - 에임이 상대한테 자동조준 거리에 상관없이 총알이 바로 꽂힘

```
Aimbot이 거리에 상관없이 총알이 바로 꽂히는 hack이므로 longestKill의 중간값 보다 월등하게 높으면서 kills의 중간값 보다 월등하게 높으면 Aimbot Hack으로 봄
```
- longestKill의 중간값 = 498.05
- kills의 중간값 = 20.5

````python
train['AimHack'] = ((train['longestKill']>800)&(train['kills']>50))
````

````python
display(train[train['AimHack'] == True].shape)
train[train['AimHack'] == True].tail()
````

## 1.6 totalDistance가 적으면서 heals,boosts,weponsAcquired 가 높은 값 = Item esp Hack

## Item esp = 모든 아이템의 위치, 거리 수집 핵

```
totalDistance의 평균값보다 극도로 이하인 값과 heals,boosts,weponsAcquired의 합의 평균보다 월등히 높으면 Item esp Hack로 봄
```
- totalDistance의 평균값 = 1764.8428507017256
- All_Items(heals,boosts,weponsAcquired의 합)의 평균값 = 6.137542540239795

````python
train['All_Items'] = train['heals'] + train['boosts'] + train['weaponsAcquired']
````

````python
train['ItemEspHack'] = ((train['totalDistance'] <= 200) & (train['All_Items'] > 10))
````

````python
display(train[train['ItemEspHack'] == True].shape)
train[train['ItemEspHack'] == True].tail()
````
