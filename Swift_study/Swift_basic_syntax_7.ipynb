# Swift Basic syntax
## 연산자

- 스워프트뿐만 아니라 다른 어떤 프로그래밍 언어를 공부하든지 연산자를 모르면 프로그래밍을 거의 할 수 없다
    - #### 그래서 가장 처음에 배우게 되는 것 중의 하나가 바로 연산자이다.
- 연산자 기호나 쓰임새는 대부분 프로그래밍 언어에서 크게 다르지 않다
    - 물론 연산자의 종류나 특징도 모두 비슷하다

### 산술 연산자
- 산술 연산자는 말 그대로 사칙 연산에 대한 연산자들이다
- 스위프트에서는 다른 프로그래밍 언어에서 제공하는 것과 거의 동일하게 다음 산술 연산자 6개를 지원한다
    - 구분 : 단항 연산자 , 연산자 : - , 사용 예 : -a , 의미 :  값의 부호를 변경
    - 구분 : 이항 연산자 , 연산자 : + , 사용 예 : a+b , 의미 : 두 개의 값을 더함
    - 구분 : 이항 연산자 , 연산자 : - , 사용 예 : a-b , 의미 : 연산자 앞의 값에서 뒤의 값을 뺌
    - 구분 : 이항 연산자 , 연산자 : * , 사용 예 : a*b , 의미 : 두 개의 값을 곱함
    - 구분 : 이항 연산자 , 연산자 : / , 사용 예 : a/b , 의미 : 연산자 앞의 값을 뒤의 값으로 나눔
    - 구분 : 이항 연산자 , 연산자 : % , 사용 예 : a%b , 의미 : 연산자 앞의 값을 뒤의 값으로 나눈 나머지
- #### 산술 연산자는 항의 개수에 따라 단항 연산자와 이항 연산자, 그리고 여기서 다루지는 않았지만 삼항 연산자 등으로 나누어진다
    - 단항 연산자는 항을 하나만 필요로 하는 것을 말한다
    - 이항 연산자는 두 개의 항을 필요로 하는 것을 말한다
        - 위의 정리에서 단항 연산자와 이항 연산자에 모두  -연산자가 포함되어 있는데, 이 둘은 서로 다른 연산자이다
        - 단한 연산자의 -는 양수를 음수로, 음수를 양수로 바꿔 주는 용도로 사용되지만 이항 연산자의 -는 연산자 양쪽의 값의 차이를 계산하는 용도로 사용
- #### 단항 연산자는 수학 용어로 바꾸자면 '부호'에 해당한다고 볼 수 있다
- #### 연산자를 사용할 때 주의해야 할 부분이 있다
    - 이항 연산자는 양쪽에 피연산자를 두는 경우가 대부분인데, 이때 피연산자와 연산자 사이의 공백은 양쪽에 동일하게 적용되어야 한다
    - 가령, 앞쪽 피연산자와 연산자 사이를 공백을 주었다면 뒤쪽 피연산자와의 사이도 공백을 주어야 한다
    - 앞쪽 피연산자와 연산자 사이를 공백없이 작성했다면 뒤쪽 피 연산자도 마찬가지이다
    - 이같이 연산자 양쪽의 공백을 일치시켜주지 않을 경우 오류가 발생할 수 있다
        - EX) 2 + 3 // = 5 (O)
        - EX) 2+4 // =6 (O) 
        - EX) 3 +6 // (X)

### 비교 연산자
- 산술 연산자는 연산 결과가 항상 다양한 값으로 나타난다
- #### 하지만 비교 연산자는 연산의 결과가 항상 True 또는 False 두 가지 중 하나로만 나타난다는 특성을 가지고 있다
    - #### 그런데 True와 False는 각각 참, 거짓을 의미하는 Bool 타입의 값이다
        - #### 결국 비교 연산자의 연산 결과는 항상 Bool 타입의 값으로 표시된다
            - #### 연산 결과가 항상 Bool 타입인 것은 논리 연산자도 마찬가지이다
- #### 비교 연산자의 종류
    - 연산자 : < , 사용 예 : a < b , 의미 : a가 b보다 작으면 true, 그렇지 않으면 false
    - 연산자 : > , 사용 예 : a > b , 의미 : a가 b보다 크면 true, 그렇지 않으면 false
    - 연산자 : <= , 사용 예 : a <= b , 의미 : a가 b보다 작거나 같으면 true, 그렇지 않으면 false
    - 연산자 : >= , 사용 예 : a >= b , 의미 : a가 b보다 크거나 같으면 true, 그렇지 않으면 false
    - 연산자 : == , 사용 예 : a == b , 의미 : a가 b와 같으면 true, 같지 않으면 false(수학에서의 = 와 같음)
    - 연산자 : != , 사용 예 : a != b , 의미 : a가 b와 같지 않으면 true, 같으면 false
- 비교 연산자는 일반적인 비교 구문에도 쓰이지만, 대부분은 논리 연산자와 함께 조건문과 분기문, 그리고 반복문을 작성할 때 사용된다.

### 논리 연산자
- #### 논리 연산자는 연산 결과를 true나 false로 표시한다는 점에서 비교 연산자와 동일하지만, 값의 크기를 비교하는 비교 연산자와는 달리 주어진 값의 논리 비교(true 또는 false)를 통해 연산을 수행한다
- 이를 위해 논리 연산자의 연산 대상인 피연산자들을 모두 true 또는 false 값을 가지는 Bool 타입이어야 한다
- #### 논리 연산자의 종류
    - 연산자 : !(NOT) , 사용 예 : !a , 의미 : a가 true이면 false, false이면 true 반환
    - 연산자 : &&(AND) , 사용 예 : a && b , 의미 : a와 b 모두 true일 때 true, 둘 중 하나라도 false이면 false 반환
    - 연산자 : || (OR) , 사용 예 : a || b , 의미 : a 또는 b 둘 중 하나라도 true라면 true, 둘 다 false 일 떄 false 반환
- 예제
    ```
    // && 연산자 테스트
    true && true // = true
    false && false // = false
    true && false // = false
    false && true // = false
    
    // || 연산자 테스트
    true || true // = true
    false || false // = false
    true || false // = true
    false || true // = true
    ```
    - 각 항이 true 또는 false일 때, 논리 연산자의 연산 결과가 어떻게 나타나는지를 보여주는 예제
    - 각 연산자와 양쪽 항에 따라 어떤 결과가 만들어지는지 확인
- #### 실무 프로그래밍에서는 논리 연산자를 비교 연산자나 산술 연산자와 함께 사용하는 경우가 대부분이다
    - 예제
        ```
        //Int 타입 상수 선언
        let a = 10
        let b = 5
        let c = 2
        
        //비교 연산자와 함께 사용하는 경우
        a > b && b > c // true && true = true
        a == b && a > b // false && true = false
        a == b || a > b // false || true = true
        
        //비교, 산술 연산자와 함께 사용하는 경우
        a - b > b - c && b == 0 // true && false = false
        a + b > c || c > 0 // true || true = true
        ```
- 주어진 논리값에 대한 논리 연산자의 처리 결과를 정리
    - a : true , b : true ,  a&&b : true , a||b : true , !a : false
    - a : true , b : false , a&&b : false , a||b : true , !a : false
    - a : false , b : true , a&&b : false , a||b : true , !a : true
    - a : false , b : false , a&&b : false , a||b : false , !a : true
- 위의 예시에서는 피연산자를 최대 두 개씩만 사용하여 논리 연산을 수행했지만, 논리 연산자는 결합법칙이 성립하므로 세 개 이상의 피연산자를 한꺼번에 논리 연산할 수도 있다
    - 이때에는 앞에서부터 두 개씩 차례대로 연산하여 나온 결과를 다시 뒤에 나오는 피연산자와 연산하면 된다
