## 연관된 데이터를 손쉽게 다루기
### 배열의 동적 선언과 초기화
- 배열을 정의할 때 한꺼번에 필요한 아이템을 몽땅 집어넣고 정적으로 정의하여 사용하는 경우보다 선언과 초기화만 해 놓은 후 필요에 따라서 그때그때 동적으로 아이템을 추가하는 경우가 훨씬 많다

#### 동적으로 배열을 정의하는 방법
- 값을 할당하지 않은 빈 배열을 선언하고 초기화할 때는 두 가지 형식을 사용할 수 있다
    - 먼저 살펴볼 형식
    
    ```
    Array <아이템 타입> ()
    ```

- #### 스위프트에서 배열을 정의하는 객체는 Array 이다.
    - #### 구조체로 정의된 이 객체는 실질적인 배열 데이터를 만드는 데 사용될 뿐만 아니라 배열에서 데이터를 효과적으로 다루기 위한 다양한 기능까지 제공

- #### 스위프트에서 배열을 정의할 때에는 반드시 저장할 아이템의 타입도 함께 명시해 주어야 한다
    - #### 컴파일러는 이때 입력받은 아이템 타입을 이용하여 다른 데이터 타입이 입력되었을 경우 잘못 입력되었음을 판단 
    - #### 또한 배열 내부의 아이템을 바로 아이템 타입 형태로 사용할 수 있도록 처리해주기도 한다

- #### 이처럼 사용 시점에서 < > 기호를 사용하여 "배열 내부"에서 "사용할 아이템 타입을 지정하는 문법"을 "Generic(제네릭)"이라고 한다.
    - 문자열 아이템을 저장할 배열이라면 <String>으로, 정수 아이템을 저장할 배열이라면 <Int>, 범용 클래스 객체를 저장할 배열이라면 <AnyObject>로 작성
- #### Generic은 구조체나 클래스 외부에서 객체 내부에 사용될 타입을 지정할 수 있다는 점에서 동적 프로그래밍 영역으로 간주
    - #### 생산성을 높여주는 문법
        - 자바나 C# 등에서 널리 사용되고 있다.

- #### 동적으로 배열을 정의할 때에는 선언과 초기화 과정이 차례로 필요하다
    - #### Declare(선언)은 "이러이러한 배열을 만들 겁니다"라고 컴파일러에 미리 알려주는 역활
    - #### initialization(초기화)는 "앞서 선언한 대로 실제로 만들어주세요"라고 요청하는 과정

- #### 객체지향 프로그래밍에서 배열 구조체는 선언만 되었을 때에는 메모리 공간을 차지하지 않다가 초기화가 진행되면 그때서야 메모리 공간을 할당받는다
    - 비로소 데이터를 저장할 수 있는 공간이 생기는 것이다
        - #### 다시 말해, 배열의 선언은 "시스템 OS에 이만이만한 공간을 사용하고 싶습니다"하고 예약하는 과정
        - #### 초기화는 실제로 사용하기 위해 체크인을 하는 과정
            - 실제로 스위프트에서 모든 변수와 상수는 알게 모르게 초기화 과정을 거친다

- 배열에서도 선언과 초기화를 동시에 처리할 수 있다

#### cities 배열을 선언하고 동시에 초기화하는 예시

```
var cities = Array<String>() // 문자열 배열의 선언 및 초기화
```

- #### 문자열 형식의 배열 객체를 정의한 다음, cities 변수에 대입
    - #### 이렇게 정의한 배열은 문자열을 저장할 수 있는 빈 배열로 초기화 된다.

#### 만약 이 배열을 단순히 선언만 하려면 타입어노테이션을 사용하여 다음과 같이 작성하면 된다

```
var cities = Array<String> // 문자열 배열을 선언
```

- #### 이렇게 선언된 배열은 초기화되지 않았으므로 아직 메모리 공간을 할당받지 않은 상태이다
    - #### 따라서 이 배열에는 아무것도 아직 저장할 수 없다.
    
    
#### 배열을 사용하기 위해서는 아래와 같이 초기화 과정을 통해 메모리 공간을 할당받아야 한다

```
cities = Array() // 배열의 초기화
```

- #### 초기화하는 과정은 비교적 단순
    - #### 배열을 정의하는 구조체인 Array 뒤에 소괄호 ( )만 붙여주면 된다
    

#### 이미 선언과 초기화가 이루어진 배열 변수를 또다시 초기화하면??
- 이때는 새로운 배열 객체가 다시 만들어져 변수에 할당된다
- 이 과정에서 기존 배열은 제거된다

#### 초기화 할 때 왜 Array<String>()이 아니라 Array()로 초기화?? 초기화할 때는 무조건 아이템 타입을 생략해도 되는 것인가??

- #### 선언이 완료된 cities 변수는 이미 타입 어노테이션을 통해서 배열의 저장 타입이 지정되어 있다
    - 필요한 것은 단지 선언된 배열을 초기화하는 구문인 것이다
        - 그래서 Array() 구문만으로 초기화해도 아무런 문제가 없다
- #### 확실하게 하려면 Array<String>()으로 초기화해도 된다.

#### cities의 선언부와 초기화 구문을 생략 없이 하나로 합치면 다음과 같다

```
var cities : Array<String> = Array<String>()
```

- 위의 선언부와 초기화 구문을 생략 없이 하나로 합친 예를 보면 무엇인가 불필요하게 중복되는 느낌이다
    - #### 이런 이유 때문에 타입 어노테이션이 작성되어 있을 경우 초기화 구문에서는 아이템 타입을 생략할 수 있도록 허용되었다고 생각하면 된다

- #### 스위프트에서 배열은 Array<아이템 타입> 으로 정의한다
    - 이 방식은 다른 프로그래밍 언어와는 다소 형식에서 차이가 있다
        - 일반적으로 프로그래밍 언어에서 배열을 선언할 때에는 자료형 뒤에 [ ] 기호를 붙이는 방법을 많이 사용한다.
            - 문자열 타입의 배열은 String[ ], 정수 타입의 배열은 Integer[ ], 실수 타입의 배열은 Double[ ], 그리고 클래스를 저장하는 배열은 MyClass[ ] 이런식으로

- 스위프트도 처음 발표되었을 당시의 배열 선언 문법은 위와 같았다
    - 하지만 몇 번의 업그레이드를 거친 후 발표된 스위프트 GM 버전에서는 String[ ]으로 선언하는 배열 문법이 제거 되었다
    - #### 대신 이와 유사한 형식으로 배열을 정의하는 구문을 제공키로 했다. 이것이 스위프트에서 배열을 정의하는 두 번째 형식이다
    
    ```
    [아이템 타입]()
    ```
    
- #### 이 구문은 대괄호 사이에 아이템 타입을 작성하는 형식으로 배열을 정의한다
    - #### 이때에는 앞에서 사용했던 배열 구조체 Array나 아이템 타입을 표현하는 < > 기호가 사용되지 않으며 단순히 대괄호 사이에 아이템 타입을 기재하는 것만으로 배열을 정의
    - 초기화할 때는 초기화 연산자인 ( )를 붙여야 한다
    - 이 구문 형식을 사용하여 배열을 정의하는 예
    
    ```
    var cities = [String]() // 배열 선언 & 초기화
    ```
    
- #### 이 구문 역시 선언과 초기화를 분리할 수 있다
    - 먼저 배열 변수를 선언할 때에는 다음과 같이 작성한다
    
    ```
    var cities : [String]
    ```
    
    - #### 초기화할 때와 거의 유사한 형태로 타입 어노테이션을 기재하되 괄호만 제외한다
        - #### 이렇게 하면 문자열을 아이템으로 갖는 배열이 선언된다
        - 단, 아직 초기화 전이므로 사용을 위해서는 초기화 과정이 필요하다
        
- #### 선언된 배열을 초기화할 때에는 다음 두 가지 형식을 사용할 수 있다
    - #### 이때 두 가지 형식이 의미하는바는 전혀 다르므로 주의해야 한다.
    - #### 배열 초기화 - 첫 번째 방식
    
    ```
    cities = [String]()
    ```
    
    - #### 배열 초기화 - 두 번째 방식
    
    ```
    cities = []
    ```

- #### 첫 번째 형식은 선언된 배열 그대로를 초기화
- #### 두 번째 형식은 빈 배열 하나를 새로 만들어서 이것을 변수에 할당한다
- #### 엄밀하게 말해서 두 번째 방식은 초기화가 아니다
    - 따라서 초기화 연산자인 ( ) 가 붙지 않을뿐더러 초기화 연산자 ( )를 붙이면 오히려 오류가 발생한다

- 이처럼 배열의 선언과 초기화 스타일은 매우 다양하다
- 필요에 따라서는 다음과 같이 선언과 초기화 표현을 서로 섞어서 사용해도 무방하다
- 어느 방식이 옳다는 것은 없으므로 익숙한 형태를 골라 사용하면 된다

#### Case 1

```
var cities : [String] // 선언
cities = [String]() // 초기화
```

#### Case 2

```
var country : [String] // 선언
country = [] // 초기화
```

#### Case 3

```
var list : [Int] = [] // 타입 어노테이션 + 초기화
```

#### Case 4

```
var rows : Array<Float> = [Float]() // 타입 어노테이션 + 제네릭 + 초기화
```

#### Case 5

```
var tables : [String]  Array() // 타입 어노테이션 + 구방식의 초기화
```

- #### 선언된 배열이 실제로 비어 있는지를 체크해야 할 경우가 있다
    - 이때는 흔히들 배열의 길이 속성인 count의 값이 0이냐 아니냐를 기준으로 하기도 한다
    - #### 하지만, 그보다는 배열 구조체에서 제공하는 속성인 isEmpty를 사용하는 것이 좋다.
        - #### 이 속성은 배열에 아이템이 없는 텅빈 상태이면 true를, 그렇지 않으면 false를 반환한다

```
var list = [String]()

if list.isEmpty {
    print("배열이 비어 있는 상태입니다.")
} else {
    print("배열에는 \(list.count)개의 아이템이 저장되어 있습니다.")
}


[실행 결과]
배열이 비어 있는 상태입니다.

```

- 이 속성은 Read-Only(읽기 전용)이므로 isEmpty 속성에 true를 대입할 수 없다
    - 설령 강제로 대입한다 해도 이로 인해 배열이 비워지지 않는다는 것을 알아두자
