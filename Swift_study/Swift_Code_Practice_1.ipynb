## Xcode 프로젝트 생성

### Xcode 템플릿 선택창 내에 템플릿 항목들이 그룹화 되어 있다
- #### 개별적인 그룹의 내용은 다음과 같다
    - #### 항목 : Application , 지원 OS : 모든 OS , 설명 : 독립 실행 프로그램, 일반적으로 앱이라고 부르는 모든 프로그램이 해당
    - #### 항목 : Framework & Library , 지원 OS : 모든 OS , 설명 : 프레임워크 또는 라이브러리, 다른 프로젝트의 개발을 돕기 위해 사용됨
    - #### 항목 : Other , 지원 OS : - , 설명 : 인앱 구매용 패키지, 또는 기타 다른 목적의 프로그램

### iOS Application 그룹 내에 Single View Application
- #### 하나의 뷰를 가지는 기본 애플리케이션을 만들어주는 템플릿
- #### 이 템플릿을 바탕으로 프로젝트를 생성하면 자동으로 만들어진 하나의 뷰를 볼 수 있다
    - #### 필요할 때마다 뷰 추가 가능

### 만들려는 앱의 형태에 따라 나머지 템플릿도 다양하게 선택 가능
- #### 그러나 처음부터 어떤 템플릿을 사용할지 명확하게 정해서 시작할 필요는 없다
    - #### 템플릿은 단지 사용을 조금 더 편리하게 도와주는 것일 뿐, 강제성이 없어서 임의로 변경할 수 있기 때문이다
- #### 실제 앱을 만들 때 기획 단계에서 화면 설계가 들어가 있으니 필요한 화면이 어떤 형식일지 알고 시작하는 경우가 대부분이다
    - #### 그러나 그렇지 않더라도 작업 도중 특정 화면 템플릿이 필요하면 그때그때 필요한 뷰를 추가하면 된다

### iOS Application의 주요 템플릿 종류
- #### 템플릿명 : Single View Application , 의미 : 하나의 뷰로 이루어진 애플리케이션. 가장 단순한 구성이며 나머지 모든 템플릿 구성의 기본이 되는 템플릿
- #### 템플릿명 : Game , 의미 : 게임 애플리케이션 제장을 위한 템플릿
- #### 템플릿명 : Augmented Reality App , 의미 : 증강 현실 기능 개발을 위한 템플릿
- #### 템플릿명 : Document Based App , 의미 : 문서 기반 기능을 구현하기 위함 템플릿, 파일 브라우저, ICloud Drive와 같은 기술을 앱 내에서 쉽게 구현할 수 있다
- #### 템플릿명 : Master - Detail Application , 의미 : 메뉴와 상세 화면으로 이루어진 복합 뷰를 기본으로 가지는 애플리케이션으로, 아이패드 또는 아이폰의 가로 모드 등 가로 사이즈가 충분할 때는 하나의 화면으로 구성되고, 협소할 때는 두 개의 화면으로 나누어 구성됨
- #### 템플릿명 : Page - Based Application , 의미 : 여러 페이지로 이루어져 차례로 쓸어넘겨 볼 수 있도록 구성되는 뷰를 가지는 애플리케이션
- #### 템플릿명 : Tabbed Application , 의미 : 탭으로 구분되는 여러 개의 뷰를 가지는 애플리케이션. 각각의 탭은 기본 뷰 컨트롤러를 가지며 탭 화면의 갯수가 일정 이상 늘어날 경우 자동으로 더 보기 메뉴를 처리해주는 특징을 가짐

### 프로젝트 정보 입력창에서의 항목들에 대한 설명
- #### 1. Product Name
    - #### 만들 프로젝트 이름을 입력하는 곳이다.
    - #### 프로젝트의 이름은 앱의 고유 식별 코드인 'Bundle Identifier' 줄여서 '번들 아이디'를 만드는 요소로 사용
- #### 2. Team
    - #### 애플 개발자 계정으로 생성된 인증서를 선택하는 부분
    - #### 앱 스토어에 빌드된 앱을 등록하거나, 혹은 인하우스 방식으로 배포할 앱을 빌드하기 위해서는 그에 맞는 인증서가 필요하다
        - #### 이때 애플 개발자 사이트에서 발급받은 인증서를 선택하여 빌드하면 목적에 맞게 서명된 앱을 만들어 낼 수 있다
- #### 3. Organization Name
    - #### 본인이 소속된 조직명을 입력하는 항목으로, 원하는 이름으로 작성하면 된다
- #### 4. Organization Identifier
    - #### 프로젝트 운영 기관 또는 조직의 식별값을 입력한다
    - #### 대체로 소속 회사나 해당 프로젝트의 도메인에서 www를 뺀 나머지를 역순으로 사용
        - #### 예를 들어 프로젝트를 진행하는 회사의 도메인이 www.google.com라면, Organization Identifier는 com.google로 작성하는 식이다
    - #### 이 값은 중요한 의미가 있는데, 앞선 Product Name 항목과 연결되어 Bundle Identifier 항목을 완성하기 때문이다
- #### 5. Bundle Idetifier
    - #### 앱의 고유 식별 코드
    - #### 이 항목은 Organization Identifier + Product Name 항목으로 자동 구성되기 때문에, 임의로 편집할 수 없다
- #### 6. Language
    - #### 앱을 제작할 개발 언어를 선택하는 항목
    - #### Objective-C와 Swift 두 가지 옵션이 있다
        - Xcode 5 버전까지는 사용할 수 있는 언어가 Objective-C 하나였으므로 이 항목이 필요 없었지만 Xcode 6 이후로는 Swift가 포함되면서 개발 언어를 선택할 수 있도록 이 항목이 추가 되었다
- #### 7. Use Core Data
    - #### 최근에는 서버에 데이터를 저장해두고 네트워크를 통해 데이터를 받아오는 방식이 일반화되어 있지만, 때에 따라서는 기기 내부의 저장소에 데이터를 저장해야 할 경우도 있다
        - #### 이럴 때 데이터 저장을 지원하는 객체인 코어 데이터를 사용할지를 선택하는 옵션
    - #### 이 옵션을 선택하고 프로젝트를 생성하면 코어 데이터를 사용하기 위한 기본 템플릿이 프로젝트에 추가되므로, 좀더 편리하다
- #### 8. Include Unit Tests
    - #### 소프트웨어의 품질 관리를 위해 최근 프로그래밍에서는 애플리케이션에 대한 반복적이고 단조로운 기능 테스트 과정을 줄이는 대신, 자동으로 테스트하고 결과를 검사해볼 수 있는 단위 테스트(Unit Tests) 기능을 제공한다
    - #### Xcode에서도 응용 애플리케이션의 기능 일부를 자동으로 테스트해주는 단위 테스트를 사용할 수 있다
    - #### 이 옵션을 선택한 상태로 프로젝트를 생성하면 'XXXTests'라는 폴더와, 기능 단위별 테스트 케이스를 직접 정의할 수 있도록 XCTestCase를 상속받은 클래스가 추가된다
- #### 9. Include UI Tests
    - #### Include Unit Tests와 비슷한 목적의 테스트 옵션이다
    - #### 다른 점은 Include Unit Tests가 기능 단위 테스트인 데 비해 이 옵션은 "화면 요소에 대한 자동 테스트"라는 점이다
    - #### 이 옵션을 선택한 상태로 프로젝트를 생성하면 XXXUITests라는 폴더와, 화면 요소를 테스트하는 스크립트를 작성할 수 있도록 XCTestCase를 상속받은 클래스가 추가된다

## 프로젝트 설정

### 프로젝트 설정
- #### 프로젝트 생성이 끝나면 프로젝트 설정 정보창이 열린다
    - #### 이 창은 다양한 설정 정보를 확인할 수 있는 공간일 뿐만 아니라 주요 속성을 설정하는 데 사용하기도 한다
    - #### 설정 창을 직접 열고 싶을 때에는 왼쪽 프로젝트 내비게이터에서 프로젝트명을 클릭하면 된다

### 설정 창은 몇 개의 영역으로 나누어져 있다. 이 중에서 특히 중요한 영역을 차례로 살펴보자
- #### [General] 탭 항목의 Identity 영역
    - #### 프로젝트 설정 정보 중에서 가장 위에 표시된 [Display Name]은 Xcode 8.0부터 신설된 항목
        - #### 모바일 기기에 설치된 앱의 이름을 설정하는 역활을 한다
        - #### 프로젝트 이름과 상관없이 원하는 앱 이름을 입력해주면 된다
        - #### 대부분 프로젝트명은 영어로 설정하지만 앱의 이름은 한글로 표시해야 할 때가 많다. 이때 이 항목을 이용하여 이름을 설정해주면 된다
- #### [Info] 탭 힝목의 Custom iOS Target Properties 역역
    - #### [Bundle Name] 이 항목의 값을 편집하면 앱 이름을 바꿀 수 있다
    - #### [Bundle Name]과 [Display Name]항목의 값이 동시에 설정될 경우 Xcode는 [Display Name]을 우선 적용한다
    - #### [Bundle Identifier]는 앱 스토어가 앱을 식별하는 고유 코드이다
        - #### 프로젝트 생성 과정에서 한번 본 적이 있는 것처엄, 프로젝트 이름과 조직명이 합해져 만들어 내는 값으로 프로젝트가 생성되고 나면 수정할 수 없다
- #### [General] 탭 항목의 Deployment Info 영역
    - #### 이 영역은 앱이 배포될 때 필요한 항목을 설정하는 부분이다
        - #### [Deployment Target]은 배포를 허용할 iOS 버저의 하한선을 지정하는 항목
            - #### 여기서 지정한 항목보다 낮은 버전의 iOS에는 설치가 제한된다
            - #### 하지만 버전을 낮게 설정한 상태에서 상위 버전을 충분히 대응하지 않으면 최신 iOS에서는 원하는 기능이 제대로 동작하지 않을 수 있으므로 주의해야 한다
        - #### [Device] 항목은 아이폰용, 아이패드용, 그리고 둘 모두를 지원하는 유니버셜 세 값 중에서 선택할 수 있다
            - #### 유니버셜로 선택할 경우 스토리보드가 자칫 매우 복잡해질 수 있으므로, 특별히 아이폰과 아이패드 양쪽을 모두 지원할 분명한 목적이 있는 경우가 아니면 아이폰이나 아이패드 중 하나를 선택하는 것이 좋다
        - #### [Main Interface]는 앱이 처음 실행될 때 기본 인터페이스 파일을 무엇으로 할 것인지 설정하는 항목이다
            - #### Main으로 선택해두면 이는 Main.storyboard 파일을 기본 인터페이스 파일로 하겠다는 의미이다
        - #### [Device Orientation]는 모바일 기기의 가로, 세로에 대한 회전 여부를 설정하는 항목이다
            - #### 각각의 방향을 결정하는 체크박스 중에서 체크된 것들에 한해 지원된다
            - #### [Device Orientation]내의 기기 방향 옵션 항목들에 대한 설명
                - #### 1. 항목명 : Portrait , 의미 : 디바이스의 기본 방향(아이폰에서는 세로 모드)
                - #### 2. 항목명 : Upside Down , 의미 : 디바이스의 위아래를 180도 회전한 상태
                - #### 3. 항목명 : Landscape Left , 의미 : 디바이스를 왼쪽으로 90도 회전시켜 눕힌 상태
                - #### 4. 항목명 : Landscape Right , 의미 : 디바이스를 오른쪽으로 90도 회전시켜 눕힌 상태
- #### [General] 탭 항목의 App Icons and Launch Images 영역
    - #### 이 영역은 기기에 설치되는 앱의 아이콘에 대한 설정과, 초기 로딩 페이지 설정을 관리한다
- #### [General] 탭 항목의 Linked Frameworks and Libraries 영역
    - #### 이 영역은 프레임워크나 라이브러리를 추가할 때 등록하는 부분이다   

## 프로젝트 구성과 스토리보드

### 프로젝트 구성과 스토리보드
- #### 프로젝트 파일의 구성 
    - #### 클래스 파일인 AppDelegate.swift와 ViewController.swift
    - #### 화면을 담당하는 Main.storyboard 파일과 LaunchScreen.storyboard 파일
    - #### 이미지 등 리소스를 관리하는 Assets.xcassets
    - #### 프로젝트의 설정을 담당하는 Info.plist 파일
    - #### 옵션 설정에 따라 XXXTests 폴더와 XXXUITests 폴더가 추가되어 있을 수도 있다

### *.swift 확장자로 이루어진 클래스 파일은 앱의 소스 코드를 구성하는 역활을 한다
- #### 기본 프로젝트에 포함되어 있는 클래스 파일은 두 개인데, 이 중에서 AppDelegate.swift는 앱 전체의 생명 주기 관리를 위임받은 객체인 앱 델리게이트를 구현한 클래스이다
    - #### 쉽게 말해 앱 전체에 적용해야 할 기능을 담당하는 클래스라고 할 수 있다
    - #### 앱이 실행되고 종료될 때, 그리고 활성화 상태가 되거나 비활성화 상태가 될 때, 백그라운드 상태로 들어가는 등의 다양한 상태 변화를 감지하고 이에 대한 처리를 해 주어야 할 때 이 클래스를 이용한다
    
### ViewContoroller.swift 파일은 뷰 컨트롤러를 구현한 클래스이다.
- #### 앱은 하나 이상의 화면을 가지는데, 이를 관리하기 위해 사용되는 것이 뷰 컨트롤러이다
- #### 일반적으로 화면의 개수만큼 뷰 컨트롤러가 필요하다

### *.storyboard 확장자로 이루어진 스토리보드 파일은 유저 인터페이스를 종합적으로 구현하는 역활을 한다
- #### 프로젝트를 생성하면 자동으로 만들어지는 스토리보드 파일은 두 개 
    - #### Main.storyboard 파일은 앱의 사용자 인터페이스 설계를 담당
    - #### LaunchScreen.storyboard 파일은 앱을 실행하면 처음 나타나는 시작 화면을 구성하는 데에 사용된다
        - #### 참고로 시작 화면은 다른 말로 "스플레시(Splash)"라고도 불린다

### 스토리보드 파일은 기존의 nib나 안드로이드의 xml 방식 화면 구성과 달리, 앱에서 사용되는 여러 화면을 하나의 파일에 모아 설계할 수 있도록 지원하는 UI 설계용 파일 형식
- #### 기존에는 앱의 화면 하나당 하나의 nib 또는 xml 파일과 대응하는 방식이었다면, 스토리보드 방식은 서로 연결되는 여러 화면이 하나의 스토리보드 파일로 구성되는 방식이라고 할 수 있다

### 인터페이스 빌더에서 스토리보드를 수정하는 작업은 쉽고 단순하다
- #### Xcode 화면의 오른쪽 아래에 있는 오브젝트 라이브러리에서 원하는 컨트롤을 골라 뷰 컨트롤러에 끌어 놓으면 된다

### 스토리보드를 이용하여 화면을 설계하는 이유
- #### 앱이 어떤 화면으로 구성되는지, 각 화면은 서로 어떻게 연관되는지, 앱의 전체 모습은 어떠한지를 한눈에 알 수 있을 것이다
- #### 하나의 스토리보드에 뷰 컨트롤러의 연결 관계를 포함하는 전체 화면 구성을 정의함으로써 앱의 전체 구조를 쉽게 파악할 수 있게 표현해 주는 것이 바로 스토리보드 방식으로 화면을 설계하는 과정이다
- #### 스토리보드를 사용하면 한 화면에서 다음 화면으로 전환하기 위해 작성해야 하는 코드를 줄일 수도 있다

### 스토리보드를 사용하면 개별 UI 파일을 이용하여 화면을 구성하는 것에 비하여 다음과 같은 이점이 있다
- #### 1. 스토리보드를 사용함으로써 앱의 화면 전체와 화면 사이 관계를 더 쉽게 파악할 수 있다. 
    - #### 스토리보드는 전체 설계가 파일 하나에 모두 들어있기 때문에 각각의 화면으로 분리된 UI 파일보다 화면의 요소들을 추적하기가 훨씬 더 쉽다
- #### 2. 스토리보드는 다양한 화면 사이의 전환을 손쉽게 처리한다. 
    - #### 이러한 전환 방식을 세그웨이(Segueway)라고 부르는데, 하나의 뷰 컨트롤러에서 전환하고자 하는 뷰 컨트롤러를 향해 <Ctrl> 키를 누른 상태로 마우스를 드래그하기만 하면 세그웨이가 자동으로 만들어진다. 세그웨이 덕분에 훨씬 더 짧은 코드로 UI를 다룰 수 있다
- #### 3. 스토리보드를 이용하면 테이블 뷰를 작업할 때 새로운 프로토타입 셀이나 정적인 셀의 외형을 만드는 것이 매우 쉽다
    - #### 인터페이스 빌더만을 사용하여 거의 완전한 수준의 테이블 뷰를 만들 수 있으며, 이렇게 작업함으로써 작성해야 하는 코드의 양이 대폭 줄어든다

### 만약에 인터페이스 빌더를 사용하는 것을 싫어하거나 전체 UI를 직접 그리는 대신 프로그래밍하듯이 코딩으로 만들고 싶다면, 스토리보드는 별 도움이 되지 않을 것이다
- #### 하지만 코드를 될 수 있는 한 짧게 작성하는 것이 유지보수에 좋다, UI에 대해서는 더욱 그렇다 때문에 스토리보드 방식의 화면 설계가 개발 생산성을 매우 높여주는 중요한 기능이라고 생각한다

## 스토리보드로 화면 구성하기

### 스토리보드로 화면 구성하기
- #### 프로젝트 내비게이터에서 Main.storyboard 파일을 클릭하여 인터페이스 빌더를 열어보자
    - #### 특별히 Xcode 버전에 차이가 없다면 인터페이스 빌더에는 한 개의 앱 UI 화면이 추가되어 있을 것이다
        - #### 이 화면을 스토리보드상의 용어로 "씬(Secne)"이라고 부른다
            - 여기서는 뷰 컨트롤러라고 부르겠다
                - #### 앱의 각 화면은 그에 대응되는 뷰 컨트롤러로 이루어지기도 하거니와, 대부분의 경우에서 뷰 컨트롤러를 추가하면 화면도 추가되기 때문이다
                - #### 따라서 지금은  '하나의 뷰 컨트롤러 =  하나의 화면'으로 이해해도 크게 무리는 없다
                    - 나중에 여러 가지 특수한 뷰 컨트롤러를 배우고 나면 '여러 개의 뷰 컨트롤러 =  하나의 화면'으로 이해가 바뀔것이다

### 추가한 뷰 컨트롤러의 화면이 아이폰에서는 한 번에 하나씩만 보이겠지만, 아이패드에서는 한번에 여러 개 화면이 동시에 보일 수도 있다
- #### Split - ViewController에서의 마스터 - 디테일 화면이 그러한 예이다

### 마우스로 뷰 컨트롤러를 클릭하면 활성화가 된다
- #### 스토리보드에 추가된 뷰 컨트롤러는 평소에는 회색으로 있다가 마우스 클릭에 의해 활성화되면 파란색 테두리가 표시되면서 현재 선택된 객체임을 나타낸다

### 뷰 컨트롤러가 잘 선택되지 않을 경우 뷰 컨트롤러의 상단에 위치한 헤더 부분을 클릭하면 뷰 컨트롤러를 정확하게 활성화할 수 있다
- #### 이 부분을 "도크(Dock)"라고 부른다

### 뷰 컨트롤러의 활성화와 관련하여 알아둘 것이 있다
- #### 객체의 속성을 설정하기 위해 사용하는 어트리뷰트 인스펙터는 항상 스토리보드에 활성화된 객체에만 적용된다
    - #### 따라서 원하는 객체가 아닌 다른 객체를 활성화한 상태에서는 속성이 적용되지 않을 뿐만 아니라 개발자가 찾는 속성 또한 올바르게 표시되지 않기도 한다
        - #### 그러니 항상 객체를 선택하여 활성화할 때는 원하는 객체가 올바르게 선택되었는지 주의해야 한다

### 뷰 컨트롤러 크기 조정
- #### 인터페이스 빌더에서 뷰 컨트롤러는 특정 모바일 기기의 크기에 맞게 사이즈를 변경할 수 있다
    - #### 이에 대한 기능을 인터페이스 하단의 View as 항목에서 제공한다
        - #### 이 항목은 인터페이스 빌더에 표시되는 뷰 컨트롤러를 어떤 사이즈로 표시할지를 설정하는 역활을 한다
- #### 사이즈 변경을 위해 View as 항목을 클릭
    - #### 아래에 숨겨진 영역이 열리면서 다양한 기기 모델이 표시된다
    - #### 왼쪽은 디바이스의 종류를 나타내고, 오른쪽은 스크린의 방향을 나타낸다
- #### 선택 항목에서 다른 디바이스 모델을 선택할 때마다 View as 레이블에는 선택된 디바이스 모델 명이 표시
    - #### 동시에 뷰 컨트롤러의 사이즈도 변경된다
    - #### 오른쪽의 Orientation을 선택하면 그에 맞춰서 뷰 컨트롤러가 서가나, 또는 눕는 것을 볼 수 있다
- #### View as 항목은 단지 현재의 인터페이스 빌더에서 뷰 컨트롤러를 어떤 크기로 보여줄지 결정하는 것에 지나지 않는다
    - #### 다시 말해 여기서 특정 디바이스 기기를 선택했다고 해서 앱이 그 디바이스에서만 동작하는 것은 아니며, 단지 화면 설계 시 편의를 위해 특정 사이즈에 미리 맞춰서 작업하는 것일 뿐이다
    - #### 또한, 여기서 가로 모드롤 설정했다고 해서 앱이 가로 모드로 동작하는 것도 아니다
        - #### 가로 모드를 위해서는 프로젝트 메타정보의 설정을 변경해 주어야 한다
- #### 그러니 이 View as 항목의 설정을 단지 편의 요소로만 이해해야 한다. 앱의 성격을 결정짓는 설정이라고 생각하면 안된다
- #### 간혹, 뷰 컨트롤러 크기를 아이폰 특정 사이즈에 맞추었다고 해도 막상 앱을 시뮬레이터로 실행해보면 원하는 결과가 나타나지 않을 수도 있다
    - #### 앱 시뮬레이터의 화면 사이즈 설정은 View as 에서 설정한 값과 별개로 동작하기 때문이다
    - #### 이는 곧 View as 항목에서 아이폰 7 사이즈로 맞추어 놓고 앱을 테스트해보더라도 무조건 아이폰 7 사이즈로 실행되지 않는다는 뜻이다
    - #### 앱 시뮬레이터를 실행시킬 때에는 어떤 버전의 아이폰 시뮬레이터를 실행시킬 것인지 지정해주어야 올바르게 실행된 화면을 볼 수 있다

### 뷰 컨트롤러에 오브젝트 추가
- #### UI 요소를 뷰 컨트롤러에 추가
    - #### 오브젝트 라이브러리에서 원하는 객체를 찾아 뷰 컨트롤러로 각각 끌어다 놓으면 된다
- #### 오브젝트 중 레이블 설명
    - #### 레이블은 원하는 텍스트를 화면에 출력해주는 객체이다
    - #### 앱 화면에 표시되는 텍스트 중 상당수가 레이블 객체를 이용하고 있을 만큼, 사용 빈도가 매우 높다
- #### 오브젝트 중 버튼 설명
    - #### 버튼은 텍스트를 화면에 표현한다는 점에서 레이블과 같지만 그에 더해 사용자와 상호 반응하면서 액션을 유도하는 역활을 한다는 특성이 있다

## 화면 전환 구현하기

### 화면 전환 구현
- #### 먼저 화면 전환을 위한 버튼을 만든다 (첫 번째 뷰 컨트롤러에)
    - #### 이제 버튼을 클릭하거나 터치했을 때 이할 화면을 추가
- #### 하나의 화면은 하나의 뷰 컨트롤러로 이루어진다
    - #### 그러니 오브젝트 라이브러리에서 View Controller 항목을 찾아 스토리보드의 빈 영역으로 드래그한다
- #### 이제 스토리보드에는 두 개의 뷰 컨트롤러가 준비되었다

### 두 개의 뷰 컨트롤러 사이에 화면 전환 기능을 구현해 보자
- #### 첫 번째 뷰 컨트롤러에 추가했던 화면 전환을 위한 버튼을 선택하고, 마우스 오른쪽 버튼을 클릭한채로 두 번째 뷰 컨트롤러로 드래그한다
    - #### <Ctrl>키와 마우스 완쪽 버튼을 같이 눌러서 드래그해도 된다
- #### 첫 번째 뷰 컨트롤러의 화면 전환을 위한 버튼에서 시작된 드래그가 두 번째 부 컨트롤러 위에서 끝나는 순간 "Action Segue" 팝업 메뉴가 나타난다
    - #### 화면 전환 방식을 선택하는 팝업이다
- #### 이 팝업 메뉴에서 [Present Modally]를 선택
    - #### 팝업 메뉴가 닫히면서 스토리보드에 표시된 두 개의 뷰 컨트롤러 사이에는 연결 화살표 하나가 추가 된다
        - #### 이 화살표 표시는 화살표가 가리키는 방형으로 화면이 전환될 수 있음을 알려주는 표시이다
        - #### 화살표 가운데 원 안에 아이폰 형식의 기호가 들어있는데, 화면 전환 방식이 달라지면 이 아이콘의 표시도 달라진다
            - #### 즉, 화살표 안에 들어 있는 아이콘은 화면 전환 방식을 표현해 주는 것이다

## 스위프트 코드 작성하기

### 스위프트 코드 작성하기
- #### 앱이 사용자의 동작에 반응하기 위해서는 레이블과 버튼 등을 제어하는 소스 코드를 작성해야 한다
- #### 이때 소스 코드를 작성하는 위치는 뷰 컨트롤러에 연결된 클래스 내부이다
    - #### 무슨 말이냐 하면, UI를 표현하는 각각의 뷰 컨트롤러에는 이를 프로그래밍적으로 제어하기 위한 클래스 객체가 배정되는데, 이 클래스의 소스 코드 내부에 스위프트 코드를 작성함으로써 UI를 마음대로 조정할 수 있다는 것이다

### 일반적으로 뷰 컨트롤러를 추가하고 나면 소스 코드를 작성할 클래스를 정의하고, 이어서 뷰 컨트롤러와 클래스를 서로 연결해 주어야 한다
- #### 하지만 기본 템플릿에 의해 만들어진 뷰 컨트롤러는 이미 ViewController 클래스와 연결되어 있는 상태이다
    - #### 이 클래스는 ViewController.swift 파일에 정의되어 있다
        - #### ViewController 클래스 내에 필요한 소스 코드를 바로 작성하면 된다
        
### 보조 에디터로 클래스 파일 열기
- #### 보조 에디터(Assiatant Editor)는 스토리보드와 소스 코드를 함께 보면서 작업하거나 서로 연결 작업을 할 수 있도록 돕는 도구이다
- #### 보조 에디터 아이콘을 클릭하면 인터페이스 빌더는 둘로 나뉜다 
    - #### 왼쪽에는 기존에 열려 있던 스토리보드 파일
    - #### 오른쪽에는 보조 에디터가 나타나면서 소스 파일
    - #### 이때 보조 에디터에 열리는 소스 파일은 대부분 현재 스토리보드에 활성화된 객체에 연결된 클래스이다
        - #### 그러니 보조 에디터를 열기 전에 먼저 첫 번째 뷰 컨트롤러를 클릭하여 확성화 해 두는 것이 좋다

### 처음 프로젝트를 생성하면서 Single View Application 템플릿을 선택하였다
- #### 이 템플릿에 의해 하나의 뷰 컨트롤러와 이에 연결되는 클래스 파일이 자동으로 생성 및 연결된다
    - #### 이 클래스 파일의 이름이 바로 ViewController.swift이다

### 반드시 이 클래스를 사용해야 하는 것은 아니다
- #### 직접 다른 이름으로 클래스를 추가하고 이를 스토리보드의 뷰 컨트롤러와 연결할 수도 있다

```
//
//  ViewController.swift
//  HelloWorld
//
//  Created by 김광준 on 23/02/2019.
//  Copyright © 2019 VincentGeranium. All rights reserved.
//

import UIKit

class ViewController: UIViewController {
    
    override func viewDidLoad() {
        super.viewDidLoad()
        // Do any additional setup after loading the view, typically from a nib.
    }    
}
```

- #### 1 ~ 7행은 클래스에 대한 주석
    - #### 주석은 소스 코드를 더 쉽게 이해할 수 있도록 덧붙이는 일종의 메모
    - #### 한 줄 주석은 // , 여러 줄에 대한 주석은 /* ~ */ 로 처리
- #### 9행은 UIKit 프레임워크를 사용하기 위해 필요한 기본 파일들을 읽어 들이는 부분
    - #### import 키워드는 라이브러리나 프레임워크 등 사용하고자 하는 기능 관련 파일을 반입하라는 뜻이다
- #### 11행 class ViewController: UIViewController {}
    - #### UIViewController 라는 클래스를 상속받아 ViewController라는 이름의 새로운 클래스를 정의하는 내용
    - #### UIViewController 클래스는 UIKit 프레임워크에 정의되어 있는 클래스
        - #### 기본 뷰 컨트롤러를 구현하는 핵심 클래스
    - #### 뷰 컨트롤러를 정의하려면 반드시 이 클래스를 상속받거나 이 클래스를 상속받은 또 다른 클래스를 상속받아야 한다

- #### 13 ~ 16 행은 viewDidLoad() 메소드를 정의하는 부분이다
    - #### viewDidLoad()는 부모 클래스인 UIVieController 클래스에 정의되어 있는 메소드로,
        - #### 뷰의 로딩이 왼료되었을 때 시스템에 의해 자동으로 호출된다
            - #### 따라서 일반적으로 리소스를 초기화하거나 초기 화면을 구성하는 등, 처음 한 번만 실행해야 하는 초기화 코드는 대부분 이 메소드 내부에 작성하면 된다

- #### 위의 ViewController 클래스에서는 부모 메소드에 정의된 이 메소드를 재정의(Override)하고 있다
    - #### 이 때문에 override 키워드가 추가되어 있다
- #### 재정의된 viewDidLoad() 메소드 내부에 작성된 super.viewDidLoad()는 부모 클래스에 정의된 viewDidLoad() 메소드의 내용도 모두 실행한다는 의미
    - #### 이 구문을 통해 상위 클래스에서 정의된 메소드 내용도 차례대로 모두 실행된다

### viewDidLoad() 메소드는 UIViewController 클래스에 구현되어 있을 뿐만 아니라, 적절한 시점에서 시스템에 의해 자동으로 호출
- #### 이른바 "콜백 메소드(CallBack Method)"

### 콜백 메소드들은 이외에도 매우 다양하다
- #### iOS 에서는 특정 시점에 처리해주어야 할 작업이 있다면 대부분의 경우 그 타이밍을 개발자가 일일히 구현할 필요 없이, 미리 정의되어 있는 메소드를 찾아 재정의하거나 필요한 코드를 추가하면 된다
