# 순서도
## Chapter 02
## 작성 방법

- #### 기본적으로 위쪽에서 아래쪽으로 처리하는 방향을 정한다
    - #### 그리고 필요한 기호를 작도한다
- #### 기호에 처리해야 하는 내용을 수식으로 적는다
    - #### 수식은 컴퓨터의 가장 기본적인 명령들 -> 입력, 기억, 산술, 관계, 논리 그리고 출력을 이용해서 작성

- 순서도의 모든 내용은 한글로 작성해도 무방하다
- #### 순서도를 이용해서 코딩 작업을 많이한다
    - #### 요사이 많이 사용되는 컴파일러들이 한글을 지원하지 않기 때문에 가급적이면 영문으로 작성

### 순서도에서 첫 번째 기호는 단말기호로 시작해야 한다
### 마지막 기호도 단말기호로 끝나야 한다
- #### 용지 가장 위쪽과 아래쪽에 단말기호를 그리고, 각각 키워드 START와 STOP를 적는다

- #### 소프트웨어를 개발하여 실행시켰을 때, 처리하고자 하는 데이터들을 기억장치에 저장해야 한다
- ### 따라서 순서도에서 처리하고자 하는 데이터들을 표현하기 위해서 대개는 시작단말기호 바로 다음은 준비기호가 작도되어야 한다
    - #### 기호상수, 변수, 배열에 대해 쉼표로 구분해서 명칭을 나열하면 된다

- ### 기호상수는 영문 대문자들로 구성되어야 하고, 반드시 "기호상수 =  값"으로 기술해야한다
    - Ex) SUBJECTCOUNT = 2.0

- ### 변수는 영문 소문자들로 구성되며, 단어나 음절을 구분할 때만 대문자를 사용한다
    - #### 초기값이 필요한 변수인 경우 "변수 = 값"으로 기술해야 한다
    - #### 여러 개의 변수들을 사용하는 경우 기호상수들 뒤에 계속해서 출력, 입력 그리고 다른 처리를 위한 변수들 순으로 쉼표로 구분하여 나열한다

- ### 입력이 있는 경우는 준비기호 바로 다음에 입력기호 한 개를 작도한다
    - #### 기억장소 한 개당 입력기호 하나씩 작도하는 것을 피한다
    - #### 한 개의 입력기호를 작도하고, 입력되는 값을 저장할 변수의 명칭을 쉼표로 구분해서 나열한다

- ### 산술연산은 처리기호에서 사용해야한다
    - #### 번호를 매기고, 총점을 구할 때 더하기 연산, 평균을 구할 때 나누기 연산은 처리기호에서 표현
    - #### 그런데 평균을 구하려면, 총점을 과목의 개수로 나누어야 한다
        - #### 과목 개수 2는 정수이다. 전산에서 정수에 대해서는 나누기 연산을 적용할 수 없다
        - ### 나누기 연산은 실수에서만 적용할 수 있다
            - #### 따라서 2가 아니라 2.0으로 실수로 나누기 표현을 해야 한다.
            - #### 그래서 과목의 개수 2에 대해 준비기호에서 기호상수로 2.0으로 표현

- ### 위쪽에서 아래쪽으로 차례대로 처리를 하도록 표현된 논리 구조를 순차구조(Sequence)라고 한다
- ### 관계 연산과 논리 연산은 비교 및 판단 기호에서 사용한다

- #### 평균 70.0을 기준으로 통과여부를 결정해야 할 때 평균이 70.0 이상인지를 표현하기 위해 비교 및 판단기호를 작도
    - average가 70.0보다 크거나 같은지 $( ≥ )$에 대한 관계식을 적어야 한다

- #### 비교 및 판단기호에 적힌 관계식을 평가해서 참인 경우 isPassed에 "PASS"를 저장
- #### 거짓이면 isPassed에 "FAIL"를 저장한다
- ### 이렇게 조건에 따라 제어 흐름이 결정되도록 하는 논리 구조를 선택구조(Selection)이라고 한다

- #### 선택구조에서 참인 쪽의 흐름선은 가능하면, 왼쪽으로 배치하고, 거짓인 쪽의 흐름선은 오른쪽에 배치
- #### 처리가 끝난 뒤에는 두 개 이상의 흐름선을 모아 하나의 흐름선을 만들도록 한다
- #### 또한 가능하면 작도하지 말아야 하지만 복잡한 경우 흐름선은 교차시켜도 무방하다
    - #### 그러나 아무런 논리적 관계를 갖지 않는다

- ### 출력기호는 끝 단말기호 바로 앞에 반드시 작도하도록 한다
    - #### 출력기호에 출력해야하는 값들이 여러 개 있는 경우, 쉼표로 구분하여 나열한다
    - ### 출력해야 하는 값들은 상수, 변수, 수식으로도 표현될 수 있다
    - ### 변수와 수식이 출력되는 것이 아니라 변수인 경우는 저장된 값이고, 수식인 경우 구해진 값이 출력된다

- #### 각 처리단계마다 처리 결과를 확인하기 위해서 즉, 디버깅에서 사용되는 출력인 에코 출력(Echo Print)은 예외

- 여기까지는 학생 한 명에 대해 성적을 처리하는 제어 논리만 표현
- 제시된 문제는 학생의 점수들이 입력되는 동안 번호를 매기고, 성적을 구하고, 성적을 출력하는 처리를 계속해야 한다
    - 이 경우 입력이 되었는지 그렇지 않은지를 결정하는 표현을 해야 한다
    - #### 따라서 비교 및 판단기호를 사용하여야 한다
    - ### 입력의 끝, 다른 말로는 파일의 끝 EOF에 대해 논리 부정을 하는 논리식을 비교 및 판단기호에 적는다
        - 논리식을 평가했을 때 참인 경우 계속하고, 거짓인 경우 끝내도록 한다

- ### 이렇게 조건에 따라 처리를 계속할지 말지를 결정하도록 하는 논리 구조를 반복구조(Iteration)이라고 한다

- #### 흐름의 방향은 위쪽에서 아래쪽으로, 왼쪽에서 오른쪽으로 통하는 것을 원칙으로 한다
    - #### 이에 반하는 경우에는 반드시 화살표를 사용한다
- ### 특히, 반복구조에서는 흐름 방향이 아래쪽에서 위쪽으로 정해져야 하는데 이때는 반드시 화살표를 이용하여 반복할지 말지를 결정하는 비교 및 판단기호의 위쪽으로 향하도록 작도되어야 한다
    - #### 그리고 그러한 흐름선은 왼쪽에 배치하도록 한다

### 주의사항
- #### 가로와 세로의 비율은 정확히 정해지지 않았으나 기호를 즉시 판별할 수 없을 만큼 바꾸어서는 안된다
- #### 추상적 표현은 사용하지 않는다. 임의의 자연수 n, 2n, 10!, 홀수냐? 와 같은 표현은 사용하지 않는다
- #### 산술연산 표현 시 단위는 사용을 하지 않는다. %, Kg등의 단위는 표시하지 않는다
- #### 정수에 대해서는 나누기 연산을 사용하지 않는다

### 정리
- 순서도는 일을 처리하는 순서를 기호들을 통해서 표현하는 도구
- 컴퓨터 기종, 프로그래밍 언어들과 무관하게 사람들의 생각을 논리적으로 표현하는데 사용할 수 있는 도구
- 이러한 도구를 이용하면 개발자들 간의 협업을 하는데 있어서 의사소통을 효율적으로 할 수 있다
- 프로그램을 개발하는 과정에서 매우 중요한 기술이 된다
