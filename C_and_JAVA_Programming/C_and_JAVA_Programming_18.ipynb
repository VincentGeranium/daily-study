# 제어구조(Control Structure)
## 반복구조(Iteration)
## 선 검사 반복구조

[p.108 순서도 참조]
### p.108에서 제시된 순서도는 선 검사 반복구조의 전형적인 구조를 나타낸다
- #### 1. 선 검사 반복구조는 반복진입 조건을 관리하는 기억장소에 대하여 초기값을 설정한다. 진입 조건을 관리하는 기억장소를 반복 제어변수라고 한다
- #### 2. 반복구조를 실행할 것인지 아니면 탈출할 것인지에 대한 조건 검사를 반복구조의 맨 처음에 수행한다. 
    - #### 만약에 반복 제어변수의 내용을 검사했을 때 조건 판단 결과가 거짓(FALSE)이 되면, 반복구조를 한 번도 수행하지 않고 반복구조를 탈출하게 된다
- #### 3. 반복 진입 조건 검사에서 결과가 참(TRUE)이면 반복구조를 실행한다.
    - #### 따라서 진입 조건 반복구조(Entry-Condition Iteration)라고도 한다
- #### 4. 반복구조의 내부에서 필요한 처리를 실행한 후 반복 구조의 맨 마지막 부분에는 항상 반복 제어변수의 내용에 변화를 준다. 반복 제어변수의 내용을 변경하지 않으면 항상 같은 값을 가지고 반복여부를 검사하므로 무한 반복에 빠지는 오류를 범하게 되므로 항상 반복 제어변수의 내용이 변경되는 부분을 반복 구조의 마지막 부분에 포함시켜야한다
- #### 5. 반복 제어변수의 내용을 판단하여 거짓(FALSE)이 되면 반복구조를 탈출하여 다음 내용을 수행한다

## 후 검사 반복구조
[p.110 순서도 참조]
### p.110에 제시된 순서도는 후 검사 반복구조의 전형적인 구조를 나타낸다
- #### 1. 후 검사 반복구조에서도 선 검사 반복구조와 마찬가지로 우선 반복 재어변수에 초기값을 설정한다
- #### 2. 후 검사 반복구조는 반복 탈출 조건의 검사를 반복구조의 맨 아래에서 실행하므로 무조건 반복구조 내로 들어와 내용을 실행한다.
    - #### 따라서 반복 제어변수의 내용 변경은 반복구조의 맨 처음에 실행하도록 한다. 후 검사 반복구조는 어떠한 경우라도 먼저 한번은 반복구조 내의 일 처리들을 실행하게 된다
- #### 3. 반복구조 내의 일 처리를 한번 실행한 후 반복 탈출 조건을 검사하게 된다.
    - #### 반복 탈출 조건을 검사해서 결과가 거짓(FALSE)이면 반복구조를 실행하고, 참(TRUE)이면 반복구조를 탈출하게 된다
    - #### 따라서 탈출 조건 반복구조(Exit-Condition Iteration)라고도 한다

### 선 검사 반복구조와 후 검사 반복구조에 대해서 알아보았다
- #### 그렇다면 두 개의 반복구조에서 어떤 구조가 과연 좋은 구조일까??
    - #### 결론적으로 선 검사 반복구조가 더 좋은 구조이다
    - #### 왜냐하면 선 검사 반복구조는 우선 반복구조의 처음에서 반복할지에 대하여 검사할 수 있으므로 조건을 만족하지 않으면 절대로 반복구조의 내용을 실행하지 않는다
    - #### 하지만 후 검사 반복구조에서는 우선 한번은 반복구조의 내용을 실행한 후에 반복구조를 탈출할지에 대하여 판단하므로 제어논리에 많은 부담을 주게 된다
    - #### 또한 이런 부담에 의하여 제어논리를 세우는데 있어서 편법을 적용할 수 밖에 없는 경우가 생기게 된다

### MAX를 10으로 조정하고 순서도를 실행해 보도록 하자
- #### 이러한 작업을 검토라고 한다
- #### 시작 단말기호를 보고 시작
- #### 다음 준비기호에서 값들을 확인
- #### MAX를 10으로 바꾸고, number는 7이고, count는 0으로 설정하도록 한다. 출력기호에 의해서 문자열 상수가 출력된다

### 비교 및 판단기호에서 number가 7이고, MAX가 10이므로 number가 MAX보다 작거나 같은지를 판단해야 하므로 7이 10보다 작으므로 참이 된다
- #### 따라서 참인 제어흐름을 따르게 된다
- #### 그래서 개수를 세게 된다. count의 값, 현재 0에다가 1을 더해서 구한 값, 1을 다시 count에 저장한다
- #### 따라서 count는 1이 된다.

### 그리고 다음 번째 7의 배수를 만들게 되는데, number의 값, 현재 7에다 7을 더하여 구한 값, 14를 number에 저장하게 된다
- #### 그리고 반복구조이므로 제어흐름을 위쪽으로 해서, 비교 및 판단기호의 내용을 처리하도록 한다
- #### 14가 MAX, 10보다 작거나 같은지를 평가해야 하므로, 14가 10보다 작지 않으므로 거짓이 되어, 제어흐름이 거짓(FALSE)쪽으로 결정
- #### 반복 처리를 끝내고 총 개수를 출력해야 한다

### 후 검사 반복구조로 작성된 순서도를 가지고 검토해 보자
[p.114 참조]
- #### 시작 단말기로호부터 시작
- #### 준비기호에서 MAX를 10으로 설정하고, number를 0, count를 0으로 설정하도록 한다.
- #### 그리고 출력기호에 의해 문자열 상수가 출력

### 다음은 7의 배수를 만든다
- #### number의 값, 현재 0에다 7을 더하여 7을 구하여 number에 저장한다. 그리고 개수를 센다
- #### 현재 count의 값, 0에다 1을 구한 값 1을 count에 저장한다
- #### 따라서 현재 count는 1이고 number는 7이 되어 있다
- #### count와 number를 출력기호로 출력하도록 하기 때문에 출력된다

### 이 부분에서 주의해야 한다
- #### 계속 처리해야 하는 내용을 한 번 처리한 후 반복을 할 것인지 판단하고 있다
- #### 즉 number가 MAX 즉, 10보다 큰 값인지 판단하도록 하고 있다
- #### number가 7이므로 10보다 크지 않으므로 거짓이 된다
- #### 선 검사 반복구조이면 평가 결과가 거짓이면 반복을 탈출하지만, 후 검사 반복구조이기 때문에 평가된 값이 거짓이기 때문에 반복 처리를 해야 한다
- #### 따라서 제어흐름은 위쪽으로 향하여, 배수를 만들고, 즉 현재 number의 값, 7에다 7을 더해서 14를 구하여 number에 저장하고, 개수를 세고, 즉 count의 값 1에다 다시 1을 더하여 2를 구하여 count에 저장
- #### 그래서 count는 2 그리고 number는 14이므로 출력기호에 의해서 출력

### 출력되지 않아야 하는 것이 출력이 되고 있다
- #### 이미 여기서 오류가 발생하는 것이다
- #### 출력이 끝난 후 반복을 계속할지 끝낼지에 대해 비교 및 판단기호에 의해서 검사하게 된다
- #### number 즉 14가 MAX 10보다 큰 것인지 판단한다
- #### 평가 결과는 참이 된다
- #### 참이므로 위쪽으로 제어흐름을 정하지 않고, 아래쪽으로 제어 흐름을 정한다
- #### 따라서 반복을 계속하지 않고, 반복을 탈출하게 되어 count를 그대로 출력하게 되면, 2가 출력이 될것이다
- #### 순서도에서 count = count - 1을 내용으로 하는 처리기호가 없이 바로 출력기호에 의해 출력이 진행된다면 10까지 7의 배수는 7뿐이므로 7의 배수가 1이어야 하는데, 2가 되어 오류가 발생하게 된다
- #### 이러한 오류를 논리오류(Logic Error)라고 한다

### 따라서 이러한 논리오류를 없애기 위해서 바로 출력할 수 없고, 현재 count에서 1을 감소시켜야 한다
- #### 따라서 반복구조를 탈출한 후에 결과를 변경해야 하는 자연스럽지 못한 제어논리가 표현되는 것이다
- #### 즉 반복구조를 탈출한 후 바로 실행되는 처리기호를 말한다
- #### 현재 count 값, 2에서 1을 빼서 1을 구한 다음에 count에 저장하도록 하고, 그 값을 바로 아래 출력기호를 이용해서 출력하도록 한다
- #### 그러면 값이 출력되고 끝 단말기호에 의해서 순서도의 실행히 끝나게 된다

### 후 검사 반복구조에서 반복을 탈출한 후 count 기억장소에 저장된 값을 1만큼 감소시킨 이유를 확인
- #### 만약에 number의 값이 100000 보다 작거나 같고, 이 값이 7의 배수라면 반복을 탈출할 수 없다
- #### 따라서 한 번 더 반복을 수행하게 되므로 실제로 반복을 탈출했을 때 10000을 넘은 7의 배수도 count를 하게 되는 문제가 발생
- #### 따라서 count의 내용을 하나 감소시켜야만 문제에서 요구하는 결과를 얻을 수 있다

### 이와 같은 문제가 발생한 이유는 무조건 한번 반복구조의 내용을 실행하므로 제어논리에 부담이 생겼기 때문에 편법으로 반복구조를 탈출한 후 하나 감소시키는 방법을 사용
- #### 선 검사 반복구조는 MAX값을 줄였을 때도 조건에 만족하지 않으면 반복을 수행하지 않는다
- #### 그러나 후 검사 반복구조는 무조건 한번은 수행한 후에 반복 탈출 조건을 검사하기 때문에 결과에 문제가 생기게 된다
- #### 이와 같이 후 검사 반복구조는 한번은 무조건 수행해야 하는 문제로 인하여 제어논리에 부담을 주게 되므로 선 검사 반복 구조를 사용하도록 한다
