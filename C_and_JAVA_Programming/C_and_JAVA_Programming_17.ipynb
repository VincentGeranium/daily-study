# 제어구조(Control Structure)
## 반복구조(Iteration)

### 순차구조와 선택구조는 단 한 번만 처리하는 경우로 이럴 때는 수작업으로 하는 것이 더 좋다
- #### 컴퓨터의 가장 큰 장점은 이런 간단한 처리라도 여러 번 처리하는데 있다
- #### 거의 모든 프로그램에서 반복처리가 필수이다
- #### 즉, 반복구조란 어떤 조건을 만족하는 동안 일정한 부분의 처리를 계속하는 구조이다

### 입력이 끝나지 않는 동안 계속해서 학생의 점수를 입력받아 학생의 성적을 평가해서 출력하여야 한다

### 먼저 한 번 입력을 받아 입력의 끝이 아닌지를 판단하여야 한다
- #### 입력이되었다면 번호를 매기고, 총점을 구하고, 평균을 구하고, 평균에 따라 통과 여부를 정하고, 결과를 출력하는 작업을 한다
- #### 그리고 다시 입력을 받아 입력의 끝이면 탈출
- #### 그렇지 않으면 성적을 평가하는 작업을 계속하는 제어구조이다

### 대부분의 문제는 순차, 선택 그리고 반복 3가지 기본구조가 섞여 있는 복합구조이다
- #### 제시된 순서도가 전형적이 예이다 [p.106 순서도 참조]
- #### 반복 구조 안에 순차구조와 선택구조가 있는 것을 볼 수 있다

### 제어 흐름(Control Flow)은 기억장소의 원리, 컴퓨터가 할 수 있는 일들 
- #### 입력, 치환, 산술 연산, 관계 연산, 논리 연산, 출력, 그리고 순차, 선택, 반복의 3가지 기본 제어 구조를 이용하여 일처리 순서를 나열하는 것이다
- #### 이와 같이 제어 흐름(Control Flow)을 구성하는 요소들 중에서 3가지 기본 제어구조는 일처리 순서와 연관하여 적용되는 요소이다

### 특히 3가지 기본 제어구조 중에서 반복구조는 전자계산기와 컴퓨터의 차이점을 나타내는 구성요소이다
- #### 컴퓨터에서는 반복을 통해서 동일한 동작을 자동으로 수행할 수 있는 장점이 있다

### 반복구조는 반복 여부 조건 검사를 반복구조의 처음에서 수행하느냐 아니면 반복구조의 마지막에서 수행하느냐에 따라서 두 가지로 나눌 수 있다
- #### 선 검사 반복구조
- #### 후 검사 반복구조

### 비유를 들어서 설명하면
- #### 배가 너무 고픈 상태에서 식당에서 밥을 먹는 상황
    - #### 식당에서 밥을 먹어야 하는데, 내 주머니에 돈이 얼마나 있는지 먼저 확인하고, 가진 돈 만큼 밥을 먹는 경우
        - #### 선 검사 반복구조
    - #### 우선 밥을 먹고 내 주머니에 돈이 얼마나 있는 지 확인하는 경우
        - #### 후 검사 반복구조
- #### 가급적이면 선 검사 반복구조를 사용하는 것이 좋은 듯 하다
