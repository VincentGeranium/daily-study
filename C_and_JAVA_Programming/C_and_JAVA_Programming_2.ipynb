# 순서도
## Chapter 02
## 토큰 (Token)

### 순서도에서 의미있게 사용되어지는 단어나 기호인 토큰들을 정리
- #### 순서도에서 각각의 역활에 따라 크게 예약어, 식별자, 연산자, 상수, 기호로 구분한다

### 예약어(Keyword)
- #### 순서도에서 이미 특별한 의미가 정해진 단어
- #### 순서도에 의해 예약된 단어
- #### 식별자로 사용할 수 없다

#### 1. AND , 논리곱 연산자
#### 2 .EOF , 파일의 끝(End of File)ㅇ에 대한 값
#### 3. FALSE , 거짓에 대한 논리 값
#### 4. NOT , 논리 부정 연산자
#### 5. OR ,  논리합 연산자
#### 6. PRINT , 출력 명령어
#### 7. READ , 입력 명령어
#### 8. START , 순서도의 시작
#### 9. STOP , 순서도의 끝
#### 10. TRUE , 참에 대한 논리 값

### 식별자(Identifier)
- #### 순서도에서 작성자에 의해 혹은 개발자에 의해서 의미가 정해지는 명칭(Name)
- #### 키워드(Keyword)는 사용할 수 없다

- ### 순서도의 경우 기억장소에 저장되는 값을 정의하기 위해 사용되는 명칭
    - #### 변수(Variable)명칭과 배열(Array)명칭 그리고 기호상수(Symbolic Constant)명칭으로 구분

### 단어해석
#### 토큰(Token)
- #### 언어의 어휘 구조(Lexical Structure)의 구성 요소로, 컴파일러(Compiler) 관점에서 의미(meanings)가 있는 최소 구성 요소

#### 기억장소(Stoarge)
- #### 식별할 수 있는 고유 주소를 가지고, 하나의 값을 저장할 수 있는 기억장치내의 특정 영역

#### 변수(Variable)
- #### 순서도 또는 프로그램 작성에서 값을 나타내는 문자나 문자들의 집합인 명칭(Name)으로 컴퓨터 프로그램의 실행 중에는 주소에 대응
    - #### 이 값은 프로그램을 실행할 때마다 또는 프로그램 내의 단계에 따라 변할 수 있는 값을 뜻함

#### 배열(Array)
- #### 같은 자료 유형의 데이터들을 저장하는 변수들의 집합으로 기억장치에 순차적으로 저장된 목록을 의미

### 상수(Constant)
- #### 순서도 또는 프로그램이 실행되는 동안에 바뀌지 않는 값을 의미한다
- #### 상수의 종류는 자료 유형에 따라 정수, 실수, 문자, 논리 그리고 문자열 상수로 구분한다
    - #### 정수 상수는 10, 0, 그리고 -55와 같이 소수점 이하 수가 없는 수를 말한다
    - #### 실수 상수는 1.0, 123.9 그리고 -11.3과 같이 소수점 이하 수가 있는 수를 말한다
    - #### 문자 상수는 'a', 그리고 'K'와 같이 영문자 한 글자를 의미한다.
        - #### 항상 작은 따옴표(' ')로 감싸서 표현한다.
        - #### 빈 칸은  (' ') 안에 스페이스 키를 눌러 한 문자를 입력해야 한다
    - #### 논리 상수는 참일 때 TRUE, 그리고 거짓일 때 FALSE
    - #### 문자열 상수는 "Dandi" , "Relly" , "Uno" 혹은 "a"와 같이 여러 개의 문자들이 모여 있는 형태를 의미한다.
        - #### 항상 큰 따옴표(" ")로 감싸서 표현한다.
        - #### 문자 상수도 큰 따옴표로 감싸서 표현하면 문자열 상수가 된다

### 기호상수(Symbolic Constant)

- #### 상수에 의미 있는 명칭을 지어, 상수에 대한 이해를 좀 더 높이기 위해 사용하는 명칭
    - #### 변할 가능성이 없는 값을 준비 단계에서 명칭을 지어 사용한다

- #### 동일한 값이라도 다른 의미이면 다른 명칭을 짓도록 한다
    - #### 순서도의 준비기호에서 가장 처음에 기술된다

#### 기호상수 예시

```
* MAX = 100
100이라는 정수 상수에 최대값이란 의미로 MAX라고 명칭을 지어 사용하도록 합니다

*PI = 3.141592
3.141592라는 실수형 상수에 PI라는 의미를 부여합니다
```

### 연산자(Operator)

- #### 주어진 값, 즉 피연산자(Operand)에 대해서 실행되어지는 평가를 규정하는 기호 또는 문자열을 말한다
- #### 쉽게 말해서 산술, 관계 그리고 논리 연산에서 연산의 종류를 나타내기 위해 사용되는 기호

#### 1. (  ) , 배열요소에 저장된 값을 읽거나 쓰고자 할 때 주어지는 위치, 즉 첨자의 배열요소의 값을 나타내는 첨자 연산자
#### 2. = , 처리기호에서는 치환 혹은 대입 연산자로 왼쪽에 적혀진 기억장소에 오른쪽에서 평가된 값을 저장한다.
#### 3. + , 처리기호에서 주어진 두 개의 값에 대해 합을 구한다
#### 4. - , 처리기호에서 주어진 두 개의 값에 대해 차를 구한다
#### 5. * , 처리기호에서 주어진 두 개의 값에 대해 곱을 구한다
#### 6. / , 처리기호에서 주어진 두 개의 값에 대해 실수 값을 구한다
#### 7. ^ , 처리기호에서 주어진 두 개의 값에 대해 제곱을 구한다
#### 8. > , 판단기호에서 왼쪽 값이 오른쪽 값보다 크면 참, 그렇지 않으면 거짓을 구한다
#### 9. ≥ , 판단기호에서 왼쪽 값이 오른쪽 값보다 크거나 같으면 참, 그렇지 않으면 거짓을 구한다
#### 10. <  , 판단기호에서 왼쪽 값이 오른쪽 값보다 작으면 참, 그렇지 않으면 거짓을 구한다
#### 11. ≤ , 판단기호에서 왼쪽 값이 오른쪽 값보다 작거나 같으면 참, 그렇지 않으면 거짓을 구한다
#### 12. = , 판단기호에서 왼쪽 값고 오른쪽 값이 같으면 참이고, 같지 않으면 거짓을 구한다
#### 13. ≠, <> , 판단기호에서 왼쪽 값과 오른쪽 값이 같지 않으면 참이고, 같으면 거짓을 구한다
#### 14. AND , 판단기호에서 두 개의 논리 값에 대해 모두 참일 때 참이고, 그렇지 않으면 거짓을 구한다
#### 15. OR , 판단기호에서 두 개의 논리 값에 대해 모두 거짓일 때 거짓이고 그렇지 않으면 그렇지 않으면 참을 구한다
#### 16. NOT , 판단기호에서 참이면 거짓으로 거짓이먄 참을 구한다

### 단어해석
#### 평가(Evaluation)
- #### 수식에서 값을 구하는 행위

### 기호(Symbol)

- #### 순서도는 원래 프로그램 작성을 위한 안내자 역활을 하기 때문에 작성자 혼자만이 이해하고, 사용자는 자신만의 전용물로만 생각해서는 안된다.
    - #### 개인적인 작업 능률을 높이는 목적 이외에 작업에 관계되는 모든 사람이 이해할 수 있고 누구나 식별할 수 있는 일정한 기호를 사용하는 것이 좋다

### 국제 표준화 기구(International Organization for Standardization)에서 추천한 기호

#### 1. 단말기호 (Terminal)
- #### 순서도의 시작과 끝을 나타낸다. 
    - #### 전산에서 처리 순서는 기본적으로 위쪽에서 아래쪽으로 그리고 왼쪽에서 오른쪽으로 진행하는 것을 기본으로 한다
    - #### 따라서 순서도에서 가장 위쪽과 가장 아래쪽에 하나씩 작도하도록 한다

#### 2. 준비기호 (Preparation)
- 우리가 사용하는 컴퓨터에서 프로그램이 처리할 데이터들은 반드시 기억장치에 저장되어 있어야 한다
    - 그리고 그렇게 저장된 데이터를 읽거나 쓰기 위해서는 반드시 식별되는 주소로 참조해야 한다
    - 컴퓨터가 식별하는 주소 단위가 아니라 주소에 대응되지만 사람들이 이해할 수 있는 의미들로 기술해서 표현하는 것이 효과적이다
    - 그래서 순서도를 전산 처리에 응용할 때는 이러한 데이터들에 대한 의미를 나타내는 명칭들을 만들어서 사용해야 한다
    - ### 이때 필요한 명칭들은 반드시 순서도의 시작 부분에 정리되어야 하는데, 이때 사용되는 것이 준비기호이다
    - ### 작업에 필요한 모든 상수, 변수 및 배열을 준비하고, 초기값을 설정할 때 사용하는 기호이다
        - #### 기억장소에 초기값을 설정할 때는 등호(=) 기호를 사용한다
    - #### 준비기호는 반드시 시작 단말기호 바로 다음에 위치해야 하고, 반드시 순서도 하나에 한 개만 작도되어야 한다.
    - #### 준비기호를 작도하고, 기호상수, 변수 그리고 배열 명칭들을 적는 작업은 이후 처리들에서 사용하겠다는 의미를 가진다
        - #### 이러한 작업을 기호상수, 변수 및 배열 선언이라고 한다

### 단어해석

#### 초기값 (Initial Value)
- #### 기억장소에 어떠한 처리에 앞서 먼저 주어져 있는 값

#### 선언 (Declaration)
- #### 프로그래밍 언어로 작성된 프로그램, 즉 원시 프로그램(Source Program)에서 그 프로그램의 번역기에 필요한 정보
    - #### 즉, 상수인지 변수인지를 알리는 작업

#### 초기화 (Initialization)
- #### 컴퓨터 프로그램에서 할당된 기억장소에 최초로 기억장소에 저장되는 의미있는 값을 설정하는 것

#### 3. 입출력 기호(Input/Output)
- #### 일반적인 입력과 출력을 나타내는 기호
    - #### 입력일 때 "READ", 출력일 때 "PRINT"를 반드시 적아야 한다

#### 4. 입력기호(천공카드, Punched Card)
- #### 외부로부터 데이터를 입력받은 경우 사용되는 기호
    - #### 천공카드(Punched Card)모양 자체가 입력을 의미하므로 "READ"를 적지 않는다
    
#### 5. 출력기호(서류, Document)
- #### 각종 원시 데이터가 기록된 서류나 종이 매체에 출력되는 결과 및 문서화된 각종 서류를 나타내는 기호
    - #### 인쇄(Print)용지 모양 자체가 출력을 의미하므로 "PRINT"를 적지 않는다

#### 6. 처리기호(Process)
- #### 치환과 산술 연산을 기술할 수 있다
    - #### 처리기호에 표현되는 연산자들은 값 복사에 대한 치환 연산자 혹은 대입 연산자 (=), 더하기(+), 뺴기(-), 곱하기(*) 그리고 나누기(/)의 산술 연산자이다
        - #### 치환연산자의 오른쪽에 적히는 변수에 저장되는 값은 절대 바뀌지 않고, 치환연산자의 왼쪽에 적히는 변수에 저장된 값은 항상 바뀐다

#### 7. 비교 및 판단기호(Decision)
- #### 주어진 조건에 따라 어떠한 처리를 해야 하는지를 결정해야 할 때 사용되는 기호이다
    - #### 관계 및 논리 연산을 해서 구한 결과가 TRUE(참) 또는 FALSE(거짓)이 된다
        - #### 이에 따라 어떠한 처리를 해야 할지를 결정할 때 사용되는 기호이다
- #### 비교 및 판단기호에 표현되는 연산자는 관계 연산자(> , ≥ , < , ≤ , = , ≠)와 논리 연산자(AND, OR, NOT)를 사용하여 수식을 만들어야 한다   

- #### 값을 비교하여 맞으면 TRUE가 되어 TRUE에 맞는 행동을 하고 그렇지 않으면 FALSE가 되어 FALSE에 맞는 행동을 선택 하는 표현구조 이러한 논리 표현 구조를 선택구조라고 한다

- #### 선택구조를 반복해서 선택하거나 몇 개의 기호들을 계속해서 처리할 것인지를 결정하는 이러한 논리 표현 구조를 반복구조라고 한다
- #### EOF는 End Of File로 파일의 끝 혹은 입력의 끝을 의미한다
    - #### 입력이 끝나지 않은 동안 순서도 내에서 반복되는 처리를 계속한다
        - #### 입력이 끝나면, 처리를 하지 않도록 하는 구조이다
- #### 기본적으로 제어흐름은 위쪽에서 아래쪽으로 그리고 왼쪽에서 오른쪽으로 결정된다
    - #### 반복구조에서는 아래쪽에서 위쪽으로 제어흐름이 결정되어 있으면, 이러한 제어 흐름을 보고, 선택구조와 구분해서 반복구조로 판단하면 된다

#### 8. 연결자 기호(Connector)
- #### 흐름선이 복잡할 경우에 원 안에 동일 숫자나 문자를 기입해서 줄구와 입구를 표시한다

#### 9. 페이지 연결자 기호(Page Connector)
- #### 순서도가 길어져서 다음 페이지에 계속 그려야 될 경우, 연결을 알려주기 위해 페이지 간의 입구와 출구를 표시한다

#### 10. 주석 기호(Annotation)
- #### 기능을 좀 더 알기 쉽도록 하기 위해 설명이 필요할 때 사용되는 기호
    - #### 기호에 적힌 내용을 누가 봐도 알기 쉽도록 설명이나 주의사항을 적어 둔다

#### 11. 흐름 방향(Flow Direction)
- #### 처리의 흐름을 표시하는데 사용되는 기호
    - #### 위쪽에서 아래쪽으로 혹은 왼쪽에서 오른쪽으로가 아닌 경우 화살표가 필요하다
    - #### 가급적이면 명확하게 하기 위해 화살표를 사용하도록 한다
