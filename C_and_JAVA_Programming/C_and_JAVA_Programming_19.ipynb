# 제어구조(Control Structure)
## 정리

### 제어논리를 표현하는데 필요한 명령어에 관련된 개념들을 학습
- #### 식과 문장, 그리고 3가지 기본 제어구조에 대하여 학습
- #### 모든 제어 논리 혹은 프로그램은 3가지 기본 제어구조를 조합하여 작성
- #### 결국 프로그램이란 기억장소의 원리와 입력, 치환, 산술연산, 관계연산, 논리연산, 출력과 3가지 기본 제어구조를 이용하여 작성하게 된다

### 3가지 기본 제어구조를 적용하는 원칙은 반드시 숙지하고 순서도를 작성

### 다음의 3가지 원칙을 기본 제어구조에 적용하도록 하자
- #### 1. 구조의 입구는 하나만 존재한다
- #### 2. 구조의 출구는 하나만 존재한다
- #### 3. 구조의 출구의 위치는 구조 맨 위나 맨 아래에 위치시켜야 한다

### 3가지 원칙을 적용함으로써 GOTO문과 같이 강제적으로 구조를 탈출하는 것을 방지할 수 있다

### * GOTO문
- #### 어떤 조건에 따라 프로그램의 정상적인 실행 순서를 바꾸어 특정 위치로 분기하는 명령어
- #### 무조건 분기 명령어라고 한다
- #### GOTO문을 남발하면 프로그램을 이해하기 어려운 코드(스파게티 코드)를 작성하게 되는 현상 발생

## 좋은 논리(혹은 프로그램) 작성 기법
### 사용자 입장에서 사용하기 쉽고 개발자 입장에서 유지 보수하기 쉽고 이해하기 쉬운 논리(혹은 프로그램)를 작성하기 위한 기법을 정리
- #### 식별자는 최대한 의미있는 명칭으로 짓도록 한다
    - 변수 명칭, 기호상수 명칭 등 개발자가 짓는 명칭은 의미가 전달될 수 있도록 짓도록 한다
    - 너무 특정한 경우만을 의미하는 명칭은 융통성이 결여되므로 일반적인 용어를 사용한다
    - 변수 명칭은 항상 소문자로 작성, 음절과 단어를 구분하는 글자만 대문자를 사용
    - 변할 가능성이 없는 내용(상수, 수식)은 의미를 부여해서 최소의 수정을 가할 수 있도록 기호상수화 한다
    - 기호상수는 처음부터 끝까지 일관성 있게 빠짐없이 적용해야 한다
    - 기호상수는 항상 대문자로 부여한다
- #### 3가지 기본 제어구조만으로 프로그램을 작성한다
    - 세 가지 기본 제어구조 정의 시 구조의 입구와 출구는 각각 하나이어야 하고 위치는 구조의 맨 위나 맨 아래에 위치시켜야 한다. 이것을 지킴으로써 강제적으로 구조를 빠져나가는(goto)방법을 봉쇄한다
    - 선택구조는 하나의 분기점에서 참(TRUE), 거짓(FALSE) 중 하나를 정할 수 있는 구조이어야 한다
    - 반복구조는 선 검사 반복구조를 시용한다. 반복구조 시작 전에 반복을 수행할지를 검사할 수 있으므로 불필요한 반복을 제거할 수 있다
    - 다중 선택문에서는 불필요한 반복을 하지 않도록 한다
- #### echo 출력을 하도록 해서, 제어논리를 추적하도록 하자. 이 때 echo 출력은 보고서 양식이 되도록, 정답을 증명할 수 있도록, 에러를 추적할 수 있도록 반복 구조 내에서 위치하도록 하자
- #### 하나의 기호나 하나의 기억장소는 한 가지 독립된 고유 기능만을 수행할 수 있도록 한다
- #### 모든 표현은 독립적이어야 한다
- #### 복잡한 표현은 한 단계씩 끊어서 설명식으로 풀어서 나열한다
    - (예)average = (koreaScore + englishScore) / SUBJECTCOUNT
        - sum = koreaScore + englishScore
        - average = sum / SUBJECTCOUNT
- #### 논리적으로 관련된 내용은 일정한 곳에 모아 둔다
- #### 반복구조 내에서는 되도록 선택구조없이 반복을 실행하는 것이 효율적이다. 선택구조 내에 반복구조를 실행하는 것은 효율성과 무관하다
- #### 누적되는 값일 경우에만 초기값을 설정해야 한다
- #### 산술 연산 중 나누기연산(/)은 실수 값을 사용한다
- #### 관계 연산 중 등가연산(=)은 반드시 정수 값에 대해서만 사용한다
