# 기억장소
## Chapter 03
## 기억장치 관리 방식

- #### 컴퓨터에 장착된 하드웨어는 한 개이고, 소프트웨어들은 동시에 여러 개 실행되는 컴퓨팅 환경에서, 기억장치는 여러 개의 소프트웨어들에 의해서 같이 사용되어야 한다
- #### 어떤 소프트웨어에 의해서 사용되고 있는 기억장소에 대해서는 다른 소프트웨어가 사용하게 되면, 사용하고 있던 소프트웨어가 제대로 동작할 수 없게 된다
- #### 따라서 사용되고 있는 기억장소에 대해서는 절대로 다른 소프트웨어에 의해서 사용될 수 없도록 하는 안전장치가 필요하게 된다
- ### 이러한 역활을 하는 소프트웨어가 운영체제이다

### 할당과 해제
- #### 응용 소프트웨어를 작성할 때 필요한 기억장소만큼 운영체제에게 사용 허가를 받아야 한다
- ### 운용체제에게 기억장소를 요청하고, 사용할 수 있는 권한을 받는 작업을 기억장소 할당(Memory Allocation, Storage Allocation)이라고 한다
    - #### 즉, 프로그램을 실행시킬 때 데이터나 프로그램 자체를 저장할 기억장소로 사용 가능케 하는 것을 말한다
- ### 그리고 사용하지 않을 때는 다시 운영체제에게 사용 허가 권한을 반납해 주어야 한다
    - ### 이러한 작업을 기억장소 할당 해제(De-allocation)라고 한다.
    - #### 그래야만 다른 소프트웨어들에 의해서 반납된 기억장소들을 사용할 수 있게 된다

- ### 언제 기억장소 할당을 하느냐애 따라 정적 할당과 동적 할당으로 구분되어 진다
    - ### 정적 할당(Static Allocation)
        - #### 일반적으로 프로그램의 실행이 시작될 때 이루어지는 기억장소 할당
        - #### 할당된 기억장소는 프로그램의 실행이 끝날 때까지 할당 해제되지 않는다
        - #### 즉 프로그램이 끝날 때 할당 해제 된다
        
    - ### 동적 할당(Dynamic Allocation)
        - #### 여러 개의 프로그램들이 동시에 실행되고 있는 시스템에서 실행 도중에 필요할 때마다 기억장치나 주변장치를 할당하는 것을 말한다
        
### 대게는 명령어와 몇 개의 데이터들에 대한 기억장소들은 정적 할당(Static Allocation)으로 관리
### 그 외 데이터들에 대한 기억장소들은 동적 할당(Dynamic Allocation)으로 관리하게 된다

### 선언과 정의
- ### 데이터를 저장할 기억장소에 대해 할당과 할당 해제를 프로그래밍에서는 변수 선언과 정의라고 한다
    - ### 선언(Declaration)이란 
        - ### 어떤 프로그래밍 언어로 작성된 원시 프로그램에서 그 프로그램의 번역기에 필요한 정보
        - ### 특히 프로그램을 실행할 때 사용될 데이터의 특성을 전달하기 위한 표현 방법을 말한다
    - ### 정의(Definition)는 
        - ### 실제로 기억장소의 크기를 지정하는 표현 방법을 말한다

- #### 몇몇 프로그래밍 언어들에서는 선언과 정의를 분리하지만, 일반적으로 선언과 정의는 동시에 진행된다
    - #### 따라서 "변수를 선언한다"라고 하면 변수를 정의하는 작업도 같이 진행되는 것으로 이해하면 된다

- ### 순서도에서는 선언과 정의를 분리할 수 없다
    - #### 따라서 변수 선언이라고 하면 정의도 동시에 진행되는 것이다
- ### 순서도에서 기억장소의 할당 표현을 하기 위해서 시작 단말기호 바로 다음에 준비기호를 작도하고, 변수 명칭들을 쉼표로 구분하여 나열해야 한다
