# 기억장소
# 기억장소의 기능

## 누적

### number = number + 1의 치환식을 검토해 보자
- #### 입력되는 순서대로 학생의 번호를 매기기 위해, 즉 1,2,3,... 식으로 하기 위해서 1씩 증가하는 표현식이다

### 치환에서 배운 것처럼, 기억장소의 원리에 따라 number에 저장된 값을 레지스터로 복사하여 저장한 다음 저장된 값에다가 1을 더하여 값을 구하고, 구한 값을 다시 numebr 주기억장치에 저장하도록 하여 1씩 증가하도록 할 수 있다
- #### 이러한 표현을 누적이라고 한다

### 얼핏 보기에 정상적으로 처리되는 것 같지만, 엄청난 오류가 발생하고 있다
- #### 할당된 기억장소에 저장되어 있는 값은 쓰레기이다
- #### 실행될 수 있는 모든 프로그램에 의해서 기억장치는 같이 사용되기 때문에 이전에 실행되었던 프로그램에 의해서 마지막으로 저장되었던 값이 존재하게 된다
- #### 따라서 쓰레기를 읽어 레지스터에 저장하고, 쓰레기에 1을 더하여 구한 값도 쓰레기일 것이다.
- #### 이렇게 구해진 쓰레기를 number에 저장해도 number에 저장된 값은 쓰레기 일 뿐이다

### 저장된 값이 쓰레기인 상태에서 number = number + 1과 같은 수식으로는 원하는 값을 구할 수 없다
- #### 원하는 값을 구하기 위해서는 number에 저장된 값인 쓰레기를 없애야 한다

### 쓰레기를 없애는 방법은 number = number + 1 수식에 의해서 값을 구하기 전에 이루어져야 하는데, 가장 간단한 방법은 초기화를 하는 것이다
- #### number = number + 1과 같이 하나의 기억장소를 이용하여 더하거나 빼는 작업을 할 수 있는 표현을 누적이라고 한다.
    - #### 전산에서 매우 중요한 개념이다
- #### 누적이란 하나의 기억장소에 저장된 값에 어떠한 값을 가감하여 자기 자신에 다시 저장시키는 것이다

### 누적에 사용된 기억장소는 할당과 동시에 최초로 자신이 원하는 값, 초기값으로 저장시켜야만 개발자가 의도하는 결과를 얻을 수 잇다
- #### 이와 같이 어떤 기억장소에 최초에 어떤 값을 저장시키는 것을 초기화라고 하고 이때 사용된 값을 초기값이라고 한다

### 초기값은 준비기호를 작도할 때 설정하지 말고, 순서도가 마무리된 후에 설정하면 쉽게 설정할 수 있다
- #### 무리하게 준비기호를 작도할 때 설정해 놓고 순서도를 작도하다 보면, 생각이 제한되기 때문에 순서도를 작도하는데 어려움을 초래할 수 있다

### 기억장소 number는 0으로 초기값을 설정한 후 1을 더하였기 때문에 반복 횟수에 따라 1,2,3,.... 순으로 값을 가지게 된다

### 그렇지만 반드시 초기화를 해야 하는 것은 아니다
- #### 누적하기 전에 반드시 유효한 값으로 설정되어 있으면 된다

### 기억장소 number에는 초기값이 필요할까?
- #### 기억장소 number에는 0을 치환하였다. 이것은 초기 값 설정, 즉 초기화라고 하지 않고 값 치환이라고 한다
- #### 이러한 경우에는 기억장소 number에는 초기값이 필요하지 않다

## 교환(Exchange)

### 기억장소의 원리에서 중요한 개념인 교환, 즉 두 개의 기억장소에 저장된 값들을 바꾸는 작업이 어떻게 이루어지는지 학습
- #### 간단하게 생각하면 아래와 같은 순서도를 작성할 것이다
- #### 그렇지만 순서도 대로 진행하면 값들을 바꿀 수 없다
- #### 우선 두 개의 기억장소, one에 1, 그리고 other에 2를 초기값으로 저장했다

```
START
↓
one = 1 , other = 2
↓
one = other
↓
other = one
↓
STOP
```

### one = other 은 기억장소 other에 저장된 값 2를 복사하여 기억장소 one에 저장하게 된다
- #### one에 저장되어 있던 값 1이 없어지고 2가 저장된다
- #### 결과는 one에 저장된 값 2는 other에 저장된 값 2와 같다
- #### other에 저장된 값은 변화가 없다
- #### 그러므로 other에 저장된 값이 one이라는 기억장소와 other라는 기억장소에 모두 저장되게 된다

### other = one는 기억장소 one에 저장된 값인 1을 기억장소 other에 복사하여 저장하도록 한다
- #### 사람의 논리로는 없어지기 전에 one에 저장되었던 값인 1이 기억장소에 저장되어 값이 교환 될 것 같으나 이것은 사람의 머리에 잔상이 있으므로 가능한 것이다
- #### 그러나 컴퓨터의 기억장소는 새로운 값으로 치환되면 이전의 값은 없어진다
- #### one = other에서 이미 one에는 other에 저장된 값이 치환되었으므로 one에 저장된 값이 other에 저장된 값과 동일하여 같은 값이 저장된다
- #### 결과적으로 one, other 모두에는 other에 저장되었던 값인 2 만이 저장되므로 교환되지 않는다

### 따라서 기억장소 두 개로는 도저히 저장된 값들을 교환할 수 없다
- #### 그래서 아래와 같은 기억장소를 한 개 더 필요로 한다
    - #### 이러한 용도의 기억장소를 임시 변수(Temporary variable)라고 한다

```
START
↓
one = 1 , other = 2 , temp
↓
temp = one
↓
one = other
↓
other = temp
↓
STOP
```

### 기억장소 one, other, 임시 변수인 temp를 할당하고, one과 other에는 각각 초기값 1,2로 저장한다
- #### temp에는 이전 프로그램에서 사용한 값, 쓰레기가 남아 있으므로 얼마인지 모른다

### 위 순서도의 실행 순서에 따라 기억장소에 저장되는 값의 변화를 추적해보자
- #### 우선 one = 1, other = 2 , temp에 대해 실행 후 내용은 다음과 같다
    - #### one에 저장된 값은 1, other에 저장된 값은 2, 그리고 temp에 저장된 값은 쓰레기이다
- #### temp = one의 의미는 기억장소 one에 저장된 값 1을 임시 변수 temp에 복사하여 저장한다는 것이다
    - #### one에 저장된 값 1이 one과 temp 두 개의 기억장소에 있다
    - #### 기억장소는 다르나 같은 값이 두 개이므로 한 개는 없어져도 상관없다
- #### one = other의 의미는 기억장소 other에 저장된 값 2를 one에 복사하여 저장한다는 것이다
    - #### other에 저장된 값 2가 other와 one 두 개의 기억장소에 있다
    - #### one에 저장된 값인 1은 2로 바뀌고 없어졌으나 temp = one에서 임시 변수 temp에 보관해 놓았다
- #### other = temp의 의미는 기억장소 other에 처음 one에 저장된 값 1을 보관해놓은 기억장소 temp에 저장된 값을 복사하여 저장한다는 것이다

### 결과적으로 one, other에 저장돤 값들이 교환되었고, temp는 교환하고자 하는 목적과는 관계가 없고, 교환 과정의 수단으로만 이용되었다

## 정리

### 결국은 프로그램에서 사용되는 값은 모두 컴퓨터 기억장치에 저장되어야 실행될 수 있다는 것을 알았다
- #### 따라서 우리가 프로그램을 잘 작성하기 위해서는 기억장소의 원리에 대하여 정확한 이해가 필요하다
- #### 기억장소는 값을 저장하는 기억장치에서의 특정 위치를 지칭한다
- #### 이와 같은 기억장소의 역활은 값을 저장하거나 저장된 값을 이용할 때 사용하는 것이다
- #### 우리가 앞으로 작성하는 모든 프로그램은 이런 기억장소의 역활을 적절하게 이용하여 원하는 결과를 만들어 내는 것이다
- #### 기억장소의 원리만 정확하게 이해하고 있다면 프로그램을 작성하는 것은 그만큼 수월해 진다는 것을 명심해야 한다

### 기억장소에 대한 이해는 아무리 강조를 해도 지나치치 않은 내용이다. 반드시 기억장소의 원리 즉, 저장시키고 이용하는 개념에 대하여 숙지하도록 하자
