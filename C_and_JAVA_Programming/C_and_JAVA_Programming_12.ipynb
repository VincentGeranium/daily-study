## 식과 연산자
- #### 연산자는 주어진 값, 즉 피연산자에 대해서 실행되어지는 평가를 규정하는 기호 또는 단어를 말한다
- #### 쉽게 말해서 산술, 관계, 논리 그리고 치환 연산에서 연산의 종류를 나타내기 위해 사용되는 기호나 단어이다
- #### 순서도에서 사용할 수 있는 연산자들은 어떠한 것들이 있는지 정리 [C_and_JAVA_Programming_2.ipynb 의 연산자 표 참조]

### 연산 방식에 따라 크게 첨자 연산자, 산술 연산자, 관계 연산자, 논리 연산자 그리고 치환 연산자로 구분한다
- #### 연산자에 따라 피연산자의 개수와 표기하는 방법을 정리하면 다음과 같다
    - #### 구분 : 첨자 연산자 , 피연산자의 개수 : 2(이항) , 표기 : 후위 , 예시 : 배열명칭(첨자)
    - #### 구분 : 산술 연산자 , 피연산자의 개수 : 2(이항) , 표기 : 중위 , 예시 : 피연산자 [+,-,*,/,^] 피연산자
    - #### 구분 : 관계 연산자 , 피연산자의 개수 : 2(이항) , 표기 : 중위 , 예시 : 피연산자 [>,≥,<,≤,=,≠,<>] 피연산자
    - #### 구분 : 논리 연산자 , 피연산자의 개수 : 2(이항) , 표기 : 중위 , 예시 : 피연산자 [AND/OR] 피연산자
    - #### 구분 : 논리 연산자 , 피연산자의 개수 : 1(단항) , 표기 : 전위 , 예시 : NOT 피연산자
    - #### 구분 : 치환 연산자 , 피연산자의 개수 : 2(이항) , 표기 : 중위 , 예시 : 피연산자 = 피연산자

### 표기법은 연산자의 위치가 피연산자를 기준으로 
- #### 앞에 놓이면, 전위(Prefix)
- #### 피연산자들 사이에 놓이면 중위(Infix)
- #### 피연산자 뒤에 놓이면 후위(Postfix)

### 예시
- #### NOT EOF , NOT 는 전위 표기법 논리 부정 연산자, EOF 는 피연산자(상수)
- #### isPassed = "PASS" , isPassed 는 피연산자(변수), = 는 중위 표기법 치환 연산자, "PASS" 는 피연산자(상수)
- #### array(1), array 는 피연산자(배열), ( )는 후위 표기법 첨자 연산자 , 1 는 피연산자(상수)

### *평가(Evaluation)
- #### 수식에서 값을 구하는 행위

## 식과 연산자
### 연산자는 평가를 하기 위해서 데이터, 즉 피연산자를 필요로 하는데 연산자에 따라 피연산자의 개수가 정해져 있다
- #### 따라서 수식을 만들 때 반드시 피연산자의 개수에 맞게 데이터를 설계해야 한다

### 사용되는 피연산자들의 개수에 따라 구분한다
- #### 단항 연산자
    - #### 피연산자를 한 개 요구하는 연산자, 논리 부정 연산자 NOT이다
    
    ```
    NOT EOF
    ```
    - #### NOT는 논리 부정 연산자이다. 피연산자는 EOF이고, 파일의 끝을 나타내는 값, 상수이다. 따라서 피연산자를 하나 사용하는 연산자이다
    - #### 논리식에 의하면 "파일의 끝이 아니면" 참이고, 파일의 끝이면 거짓이 된다

- #### 이항 연산자
    - #### 피연산자를 두 개 요구하는 연산자이다. 대부분의 산술, 관계 그리고 논리 부정 연산자를 제외한 논리, 그리고 치환 연산자는 이항 연산자이다
    
    ```
    isPassed = "PASS"
    ```
    - #### isPassed는 피연산자(변수), = 는 치환 연산자 , "PASS"는 피연산자(상수)
    - #### 치환 연산자는 두 개의 피연산자를 갖는 이항 연산자이다
    - #### 치환 연산자는 등호 기호를 사용한다
    - #### 등호 기호의 왼쪽에 있는 피연산자는 변수이고, 오른쪽에 있는 피연산자는 상수이다
    - #### 치환 연산자는 오른쪽에 표현되어진 값을 주기억장치에 할당된 기억장소에 값을 복사해서 저장하는 연산이다
    - #### 따라서 치환연산자에서 사용되는 피연산자 중에서 왼쪽에 사용되는 피연산자는 반드시 변수이어야 한다

## 식과 피연산자 형식
### 식에 사용되는 피연산자는 상수, 변수만이 사용되는 것이 아니라 수식도 사용될 수 있다
- #### 번호를 매기기 위해서 사용된 식은 치환식 관점에서 보면, number + 1 산술식이 피연산자로 사용된 것이다

```
number = number + 1
```

- #### number = number + 1 은 치환식 , number + 1은 산술식

### 총점을 구하는 식, 평균을 구하는 식 등도 피연산자로 식을 사용하는 식들이다

```
sum = koreanScore + englishScore
average = sum/SUBJECTCOUNT
```

## 평가 순서
### 식을 만드는데 피연산자로 식이 사용되어진 식을 복합 수식(Compound Expression)이라고 한다
- #### 이때는 어떠한 식이 먼저 평가되어지는지에 따라 다른 값을 구할 수 있다

### 따라서 정확한 값을 구하기 위해서는 수식을 평가하는 순서에 대한 개념들이 정리되어 있어야 한다
- #### 우선순위(Precedence)
- #### 결합성(Associativity)

### 우선순위와 결합성에 대한 구분과 기호
- #### 구분 : 첨자 연산자 , 기호 : ( ) , 우선순위 : 1 , 결합성 : →
- #### 구분 : 논리 연산자 , 기호 : NOT , 우선순위 : 2 , 결합성 : ←
- #### 구분 : 산술 연산자 , 기호 : ^ , 우선순위 : 3 , 결합성 : →
- #### 구분 : 산술 연산자 , 기호 : * , / , 우선순위 : 4 , 결합성 : →
- #### 구분 : 산술 연산자 , 기호 : + , - , 우선순위 : 5 , 결합성 : →
- #### 구분 : 관계 연산자 , 기호 : >, ≥, <, ≤, =, ≠ 혹은 <> ,우선순위 : 6 ,결합성 : →
- #### 구분 : 논리 연산자 , 기호 : AND , OR , 우선순위 : 7 , 결합성 : →
- #### 구분 : 치환 연산자 , 기호 : = , 우선순위 : 8 , 결합성 : ←

### 우선순위는 애매모호하게 해석 가능한 수식에서 어느 연산을 먼저 계산할 것인가를 결정하는 규칙이다
- #### 일반적으로는 소괄호 ( ) 로 싸여진 수식, 산술식, 관계식, 논리식 그리고 치환식으로 평가 순서가 정해진다

### 같은 우선순위를 갖는 수식으로 구성된 복합수식에서는 결합성을 적용해야 한다
- #### 결합성은 우선순위가 같은 경우 수식을 어느 방향에서부터 먼저 계산할 것인가를 정해 놓은 것이다

### 왼쪽에서 오른쪽으로의 결합 순서를 갖는다는 것은 아래와 같이 계산되는 식을 말한다
- #### (1 + 2) 수식을 먼저 평가하고 평가된 값 3과 3간의 더하기 식을 평가하는 순서로 수식을 평가한다

```
(((1 + 2) + 3) + 4)
```

### 역으로 오른쪽에서 왼쪽으로 결합 순서를 갖는다는 것은 아래와 같이 계산되는 식을 말한다
- #### (3 + 4) 식을 평가하고, 평가된 값 7을 가지고, (7 + 2) 식을 평가하는 순서로 평가한다

```
(1 + (2 + (3 + 4)))
```

### 산술식, 관계식 그리고 논리식에서는 왼쪽에서 오른쪽으로, 치환식에 해서는 오른쪽에서 왼쪽으로 평가 순서가 정해진

### 번호를 매기는 처리기호에 작성된 식 number = number + 1은 산술식과 치환식으로 구성되는 복합 수식이다
- #### 이러한 경우에는 정확한 값을 구하기 위해서 우선순위에 의해서 산술식을 먼저 평가
- #### 산술식에 의해서 구해진 값을 가지고 치환식을 평가해야 한다
- #### 즉 산술식에 의해서 구해진 값을 number에 복사하여 저장하도록 한다

```
number = number + 1

평가 순서 
(1) 산술식 : number + 1
(2) 치환식 : number = (1)산술식에 의해서 구해진 값
```

### AND 논리곱 연산자에서는 두 개의 피연산자의 값이 모두 참일 때만 참이기 때문에 왼쪽항의 값이 거짓이면, 논리식 이후의 관계식을 평가하지 않고 바로 거짓으로 평가하게 된다

```
1 > 2 AND 3 < 4

평가 순서:
(1) 관계식 : 1 > 2    : FALSE
(2) 관계식 : 3 < 4    : TRUE
(3) 논리식 : FALSE AND TRUE    : FALSE
```

### 위 식은 두 개의 관계식과 한 개의 논리식으로 구성된 복합 수식이다
- #### 이 식을 평가하는 순서에 대해 알아보자
    - #### 우선순위에 의해서 관계식을 논리식보다 먼저 평가해야 한다
    - #### 그리고 두 개의 관계식에 대해서 결합성으로 왼쪽에 있는 관계식을 먼저 평가하고 오른쪽 관계식을 평가해야 한다
    - #### 따라서 1 > 2 관계식을 제일 먼저 평가하게 된다
    - #### 그리고 두 번째로 3 < 4 관계식을 평가하게 된다
    - #### 마지막으로 두 개의 관계식을 평가한 결과들로 논리식을 평가하게 된다
    
### 그렇지만 위에서 이야기한 것처럼 논리곱 연산자는 제시되는 두 개의 연산자가 참일 때만 참이기 때문에 먼저 제시된 피연산자의 값이 거짓이면, 이미 논리곱 연산자에 의해서 구해지는 값은 거짓
- #### 때문에 뒤에 제시되는 피연산자의 값을 평가하지 않는다
- #### 1 > 2 AND 3 < 4 복합수식에서는 실제로 첫 번째 식 1 > 2를 평가한 다음은 식 평가가 끝나게 된다
- #### 즉 3 < 4 식도, 논리식도 평가하지 않게 된다

### 똑같은 원리로 OR 논리곱 연산자에서는 두 개의 피연산자의 값 중에서 하나라도 참이면 참이기 때문에 왼쪽의 값이 참이면, 논리식 이후의 관계식을 평가하지 않고 바로 참으로 평가하게 된다

### 논리곱 연산자나 논리합 연산자에서 왼쪽 값만으로 TRUE, FALSE를 판별 가능한 경우에는 오른쪽 값에 대해서는 평가하지 않는다
- #### 이러한 현상을 Short circuit rule이라고 한다
- #### 평가하는 횟수가 줄어드니까 실행속도도 빨라지게 된다

### 우선순위와 결합성에 따르지 않고, 먼저 해야 하는 식이 있다면 소괄호 ( )를 이용하여 감싸서 나타내면, 소괄호로 싸여진 식이 가장 먼저 평가되어 진다

## 식과 자료유형
### 하나의 연산자에 사용되는 피연산자들의 자료유형은 같아야 한다
- #### 그렇지 않으면 구하는 값이 정확하다고 보장할 수 없다
- #### 따라서 같은 자료형을 갖는 피연산자를 갖고 연산을 해서 같은 자료형의 값을 구하여야 한다

```
average = sum / SUBJECTCOUNT
```

### 위 식은 평균을 구하기 위한 식이다
- #### sum은 정수형이고, average는 실수형이다 그리고 평균을 구할 때 사용하는 과목의 개수는 정수형으로 2이어야 한다
- #### 위 식에서 오른쪽 식 sum / SUBJECTCOUNT에서 SUBJECTCOUNT는 2이어야 한다
- #### 사람들이 생각하는 수식이면 문제가 없다
- #### 나누기를 하면 당연히 실수형 값이 구해진다고 생각하기 때문이다
- #### 그렇지만 컴퓨터에서는 정수형 데이터에 대해서는 나누기 연산을 사용할 수 없다
- #### 정수형인 피연산자들을 가지고 나누기 연산을 하면, 실수형 값을 구하는 것이 아니라, 정수형 값을 구하게 된다. 소수점 이하를 버리게 된다

### 따라서 원하는 평균을 구하기 위해서는 나누기에 의해서 구해지는 값이 실수형이어야 한다
- #### 따라서 sum은 입력되는 값들이 정수형이어서 실수형으로 할 수 없고, SUBJECTCOUNT를 실수형으로 하여야만 실수형 값을 구할 수 있게 된다
- #### 따라서 준비기호를 보면 SUBJECTCOUNT의 값을 2로 하지 않고, 2.0으로 실수형 상수를 표현하고 있다 [p.89 순서도 참고]

### 이렇게 정수형을 실수형으로 강제적으로 변환하는 작업을 형 변환 (Type Casting)이라고 한다
- #### 물론 특정 프로그래밍 언어에서는 이러한 작업도 연산자를 이용하여서 하게 된다
- #### 그러나 순서도에서는 연산자를 제공하지 않기 때문에 상수를 이용해서 형 변환에 대해 표현해야 한다

### 또한 식을 만드는데 자료형에 대해 주의할 것은 실수형 피연산자를 가지고 같은지(=)와 같지 않은지(≠)에 대한 등가 비교를 할 수 없다
- #### 왜냐하면 정밀도, 즉 소수점 이하 자리수에 대한 표현에 대해 정확성을 보장할 수 없기 때문이다
