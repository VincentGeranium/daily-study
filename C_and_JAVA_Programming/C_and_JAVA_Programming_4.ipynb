# 기억장소
## Chapter 03
## 컴퓨터의 정보 처리 단위 : Bit, Byte 그리고 Word

### 비트, 바이트 그리고 워드
- #### 현재 사용되고 있는 컴퓨터들은 어떤 일을 처리하는데 필요한 모든 명령어와 관련 데이터들을 주기억장치에 둔다
    - #### 주기억장치에 있는 모든 명령어와 관련 데이터들을 중앙처리장치에 의해서 명령어를 하나씩 꺼내어 실행시키는 방식이다
    
- 
```
 따라서 컴퓨터가 어떠한 일을 처리하는데 있어 필요한 명령어와 관련 데이터들을 어떤식으로 기억장치에 표현하고,
 이들을 처리하기 위해서 주기억장치에 어떻게 쓰고 읽어내는지에 대한 이해는 소프트웨어 개발자에게 매우 중요한 것이다
```

### 단어해석
#### 명령어 (Instruction)
- #### 기계 명령어라고도 하고, 중앙처리장치의 제어 논리(Control Logic) 장치가 이해할 수 있는 0과 1로 조합된 단어

### 주기억장치의 용량
- #### 모든 컴퓨터는 주기억장치로 사용되는 램(Randon Access Memory, RAM)을 내장하고 있다
- #### 컴퓨터 기억장치의 용량은 일반적으로 킬로바이트(KB : KiloByte) 또는 메가바이트(MB : MegaByte)로 표시
    - 1 KB = 1,024 Byte
    - 1 MB = 1,048,576 Byte

- 이것이 어느 정도의 크기인지 가늠하기가 어려울 것이다
    - #### 크기를 가늠하기 위해서는 세는데 있어 기준이 되는 가장 작은 기준 단위로 표현해 보아야 한다
- ### 컴퓨터에서 데이터 혹은 정보를 처리하는데 가장 작은 기준 단위가 비트(bit)
    - #### 1 Byte = 8 Bit이다

### 비트 (Bit)
- #### 1,879,048,192 개수처럼 주기억장치는 보통 수십억 혹은 수천억에 달하는 많은 기억소자(Storage Cell)들로 구성된다
    - #### 그들 각각은 0 혹은 1의 값을 가지는 즉, 한 자리의 이진수를 저장할 수 있다
- ### 컴퓨터가 정보를 표현하는 최소 단위로 비트(Bit, Binary Digit)라고 한다
- #### 이처럼 한 개의 비트는 극소량의 정보 밖에는 표현할 수 없기 때문에 비트 개개로는 거의 정보가 처리되지 않는다
- #### 통상 일정한 크기의 그룹들의 단위로 하여 처리되고 있다

### 단어해석
#### 기억소자 (Storage Cell)
- #### 기억장치를 구성하는 단위

### 바이트 (Byte)
- 사람들이 컴퓨터를 개발한 이유는 데이터 혹은 정보를 처리하기 위해서이다
    - #### 따라서 컴퓨터를 사용하여 정보 처리를 하기 위해서는 사람들간의 의사소통을 하기 위해 사용되는 숫자, 문자, 기호를 컴퓨터에 표현해야한다
        - #### 그러기 위해서는 몇 개의 비트를 조합하여야 한다
            - ### 8개의 비트를 조합하면 0에서 9까지 숫자, A에서 z까지 영문자 그리고 특수기호를 표현할 수 있다
- ### 8개의 비트가 묶어진 정보 처리 단위를 바이트 (Byte)라고 한다
    - #### 컴퓨터와 사람간의 정보를 교환하는 단위로 컴퓨터가 처리하는 최소한의 입출력 단위이다

### 워드 (Word)
- #### 정보를 처리하기 위해 컴퓨터의 기본 기능 장치들 -> 입력장치, 기억장치, 연산장치, 제어장치 그리고 출력장치 간의 데이터들이 전송되어야 한다
    - #### 이때도 비트 단위로 전송되는 것이 아니라, 8, 16, 32, 64개의 비트의 그룹 단위로 정보가 저장되고 읽혀질 수 있도록 조직되어 있다
- ### 즉 정보가 처리되기 위해서는 명령어와 데이터들이 주기억장치로부터 읽어져서, 중앙처리장치로 복사되어야 한다 
    - #### 이렇게 한 번 읽어져서 복사되어지는 비트의 그룹 단위를 워드(Word)라고 한다
- ### 워드(Word)를 구성하는 비트들의 개수를 워드 크기(Word Length)라고 한다
- ### 컴퓨터의 경우 중앙처리장치에 있는 기억장치인 레지스터(Register)의 크기와 운영체제에 따라 나뉜다
    - ### 8비트, 16비트, 32비트이거나 최근에는 64비트의 워드 크기(Word Length)를 가진다

### 단어해석
#### 레지스터 (Register)
- #### 산술 및 논리 연산이나 정보 해석, 전송 등을 할 수 있는 일정 길이의 정보를 저장하는 중앙 처리 장치(CPU)내의 기억 장치
