# 순서도
## Chapter 02

- 순서도는 일처리 간소화와 작업 합리화를 위해 우리 사회에서 많이 사용되고 있다.
- 특히, 컴퓨터에 의해 어떤 업무를 처리하기 위해서는 없어서는 안 될 만큼 중요하다.
    - 프로그램을 작성하는데 있어서 문제 해결을 위한 논리적인 순서만 완전히 수립되면, 프로그래밍 언어를 이용하여 코딩하는 등의 나머지 작업은 쉽게 할 수 있다
    - 또한, 프로그램의 형태는 컴퓨터의 기종이나 프로그래밍 언어에 따라 다르지만 이것을 유도하는 논리적인 방식에는 차이가 없다
        - 그러므로 완성된 순서도는 모든 컴퓨터에 적용될 수 있다
- 순서도가 프로그램을 개발하는 작업에 있어서 중요한 도구가 된다
### 프로그램을 작성하는 경우 무엇보다도 처리의 순서나 흐름이 중요하므로 순서도를 "얼마나 잘 작성하느냐"하는 것은 곧 "얼마나 우수한 프로그램이 되느냐"하는 문제 해결의 열쇠이다.

### 프로그램을 만든다는 것은 문제를 해결하는 것이다
- #### 새로운 명령어를 만드는 것이 아니다
- #### 컴퓨터가 제공하는 명령어 목록에서 문제 해결에 필요한 명령어들을 선택해서 실행 순서 및 횟수를 결정한다
    - #### 각 명령어가 실행되는데 필요한 데이터들을 정확하게 정해주는 작업을 하면 되는 것이다.

### 컴퓨터가 할 수 있는 일, 즉 컴퓨터의 기본 기능들은 입력, 기억, 연산, 출력 그리고 제어이다
- #### 처리할 문제에 대해 어떠한 컴퓨터의 기본 기능들을 사용할 것이고, 언제 어디서 그리고 몇 번 실행하도록 할 것인지를 결정해야 한다
- #### 또한 결정된 컴퓨터의 기본 기능들에서 데이터들이 사용되는 경우, 어떠한 데이터들이 사용될 것인지를 결정해 주어야 한다

### 말로 혹은 글자로 표현하는 것보다 순서도로 누구나 이해하고 있는 표준화된 기호들로 직관적으로 나타내는 것이 더욱더 효과적인 방법이다

## 정의
- #### 컴퓨터로 처리하고자 하는 문제를 이해하여 그 처리 순서를 단계화
- #### 그리고 처리 단계 상호 간의 관계를 일정한 기호를 사용하여 일처리 순서를 일목요연하게 나타낸 그림
### 순서도를 이용하면 개발자 상호간의 의사 전달을 명확하게 해주고, 일처리 과정에 오류가 있거나 변화에 대해서 쉽게 고칠 수 있다

## 특징
- #### 논리적인 체계를 쉽게 이해할 수 있으므로, 업무의 전체적인 체계를 일목요연하게 파악 할 수 있다
- #### 프로그램의 설계도로 이용할 수 있어 순서도에 따라 프로그램을 코딩한다

- 전산 처리되는 업무는 대개가 복잡하여 하루 이틀에 프로그램을 작성할 수 없다
    - 이렇게 복잡한 업무를 순서도를 작성하지 않고 직접 프로그래밍 언어를 이용하여 코딩한다는 것은 무리가 따른다.
- 설사 코딩이 이루어진다 해도 정확성이 없고 최대한의 능력을 발휘할 수도 없다
- 이것은 초안 없이 원고를 작성하는 것과 같이 불안이 따르기 때문에 반드시 순서도에 의한 코딩이 이루어져야 한다

- #### 프로그램의 흐름에 대한 추적이 용이해서 문제 해결의 정확성 여부를 쉽게 판단할 수 있다

- 순서도는 논리적인 체계를 그림으로 나타낸 것이기 때문에 프로그래밍 언어의 규칙에 따라 코딩된 프로그램보다 정확성이 높고, 또 이해하기 빠르다
    - 따라서, 테스트할 때 오류가 발생한 경우 순서도를 검토하여 확인해서 고치는 것이 업무 처리에 있어 효율적이다.

- #### 프로그램을 유지보수(Maintenance)할 때 사용할 수 있다

- 이미 개발된 시스템을 잘 운영하기 위해서는 프로그램의 보존이 필수적으로 따라야 한다
- 이때 암호화된 프로그램만을 보관하는 것은 원본이 없는 서류를 보관하는 것과 흡사하다
    - 그러므로 언어나 기종에 구애받지 않는 순서도를 보관하여 어떠한 경우에도 완전하도록 대처해야 한다
        - 뿐만 아니라 기능적인 보강이나 시스템의 성능 개선 작업에 바로 활용 할 수 있다
