# 기억장소
# 기억장소의 기능

### 할당된 기억장소에 저장된 값이 어떤 것인지 알 수 없다
- #### 왜냐하면 이전에 실행되었던 프로그램에서 무엇을 저장시켰는지 알 수 없기 때문이다
- #### 기억장소에 저장된 값은 새로운 값이 저장될 때까지 없어지지 않기 때문이다
- #### 각각의 기억장소에 저장된 값은 이전 프로그램에 의해서 마지막으로 저장된 값이다
    - #### 이것은 실행되어진 프로그램에 유효한 값은 아닐 것이다
        - #### 따라서 이러한 값을 쓰레기(Garbage)라고 한다

### 이러한 쓰레기(Garbage)를 가지고 있는 할당된 기억장소를 곧바로 수식에 사용하면, 원하지 않는 결과, 즉 논리 오류가 발생한다
- #### 논리 오류가 발생하는 것을 방지하기 위해서 할당된 기억장소를 이후 처리들에서 사용하기 위해서는 할당된 기억장소에 합당한 값을 저장시켜야 한다
    - #### 할당된 기억장소에 값을 저장하는 방법들로는 초기화(Initialization), 입력(Input), 그리고 치환(혹은 대입 Assigment)이 있다

## 초기화

### 기억장소의 할당과 동시에 할당된 기억장소에 최초로 원하는 값을 저장해서 쓰레기를 치우는 작업을 초기화라고 한다
- #### 기억장소가 할당될 때 기억장소에 최초로 저장되는 값을 초기값(Initial Value)이라 한다
- #### 이렇게 최초로 할당된 기억장소에는 자신이 원하는 값을 저장시켜야만 의도하는 결과를 얻을 수 있다
    - #### 특히 누적에 사용되는 기억장소에 대해서는 반드시 초기화를 해야 한다

### 초기화는 준비기호에서 초기화해야 하는 기억장소에 대한 변수 명칭 뒤에 등호를 적고 초기값을 적으면 된다

[p.74  순서도 참조]
- #### number에 대해 초기값 0으로 초기화를 하고 있다
    - #### 왜냐하면 기억장소 number 하나를 이용해서 수를 1씩 증가시키는 표현을 하고 있기 때문이다
        - #### 이러한 표현은 누적이라고 한다
- #### 따라서 number는 초기화가 되어야 한다

### 초기값은 0 아니면 1 중에 어느 하나를 임의로 결정하는 것이 아니라, 첫 번째 수식이 평가 되었을 때 값을 정하고, 수식을 정리하면 초기값이 구해진다

[p.74  순서도 참조]
- #### number = number + 1에 의해서 첫 번째로 구해져야 하는 값은 1이어야 한다
    - #### 치환 연산자의 왼쪽에 적힌 number에 저장되는 값이 1이어야 한다는 것이다

### 따라서 수식을 정리해 보면 다음과 같다
- #### 1 = number + 1 이 수식에 대해 정리하면 다음과 같다
    - #### number = 0 따라서 수식을 정리하여 구한 값이 초기값이어야 하기 때문에 초기값은 0이어야 한다

## 입력 (Input)

### 입력장치를 통하여 입력된 데이터를 저장하는 방식이다
- #### 학생 점수, 상품 금액 그리고 은행 입출금액 등과 같이 소프트웨어 개발자가 마음대로 정할 수 없는 값으로 사용자에 의해서 정해지는 값이다
- #### 외부 입력매체로부터 읽어 들여 저장하는 방식이다
- #### 입력으로 값을 저장하는 기억장소에 대해서는 반드시 초기화할 필요가 없다
    - #### 입력을 하지 않은 상태에서 이용하지 않기 때문이다

### 기억장소들 name, koreanScore, englishScore에 저장된 값은 모두 쓰레기이다
[p.75 순서도 참조]
- #### 이러한 기억장소를 수식에 이용하면 원하지 않는 결과를 얻을 수 있다
    - #### 그래서 입력에서 사용되는 기억장소들, name, koreanScore, englishScore는 입력을 받아 이후 처리들에서 사용하여야 한다

## 치환 혹은 대입 (Assigment)

### 상수, 변수 그리고 수식으로 구해진 값으로 기억장소에 값을 저장하는 방식이다
[p.77 순서도 참조]

### isPassed에 저장된 값은 아직 이전 프로그램에 의해 사용됐던 값으로 isPassed에 저장될 값으로는 유요한 "PASS"와 "FAIL"이 아닌 어떠한 값일 것이다
- #### 이러한 값을 쓰레기라고 한다
- #### 이러한 쓰레기를 없애는 방법으로 치환 기능을 이용한다

### isPassed = "PASS", isPassed = "FAIL"에서 isPassed에 실행해서도 변하지 않는 값, 즉 문자열 상수, "PASS"와 "FAIL"를 isPassed에 저장하는 치환 표현이다

### sum = koreanScore + englishScore는 두 개의 변수, koreanScore와 englishScore에 저장된 값들을 읽어 중앙처리장치로 복사하고, 중앙처리장치에서 두 개의 값을 더하여 값을 구한다
- #### 이렇게 구해진 값을 sum에 저장하라는 표현이다
- #### 산술식으로 구해진 결과를 특정 기억장소에 값으로 저장하여 바꾸는 치환 표현이다

### average = sum / SUBJECTCOUNT는 sum = koreanScore + englishScore에 의해서 방금 저장된 합을 다시 중앙처리장치로 읽어와서 SUBJECTCOUNT 즉 2.0 상수로 나누어 값을 구한다
- #### 이렇게 구해진 값을 average 라는 기억장소의 값으로 저장하는 치환 표현이다

### 등호의 우측에 있는 값을 좌측에 있는 기억장소(변수)에 저장하는 기능을 치환 혹은 대입 기능이라고 한다
- #### 이때 등호기호(=)를 사용하고, 이것을 치환연산자 혹은 대입연산자라고 한다
- #### 치환연산자를 사용한 표현식을 치환식 혹은 대입식이라고 한다

### 치환식에서 등호( = )를 기준으로 왼편은 왼쪽 값(L-Value), 즉 주기억장치에 저장된 값, 
### 오른편은 오른쪽 값(R-Value), 즉 레지스터에 저장된 값이라고 한다

- #### 왼쪽 값(L-Value)에는 항상 변수만 사용할 수 있다는 점을 명심
    - #### 주기억장치에 있는 값을 말하는 것이다
- #### 오른쪽 값(R-Value)은 주기억장치에 있는 값을 복사해서 중앙처리장치에 있는 레지스터(Register)로 저장한 값을 말한다

### 따라서 주기억장치의 값을 변경하지 않는다는 것을 명심
- #### 원래 값을 이용만 한다는 점을 명심
- #### 오른쪽 값으로는 상수, 변수 그리고 수식으로 표현될 수 있는데 실제는 값이다

## 데이터 처리
- #### 명령어를 실행하기 위해서, 먼저 주기억장치의 명령어가 중앙처리장치로 전달되어야 하며, 이 작업은 중앙처리장치 내의 제어 회로에 의해 이루어진다
- #### 그리고 처리할 값들과 처리된 결과 값 또한 주기억장치와 중앙처리장치 사이에서 이동되어야 한다
- #### 그렇기 때문에 주기억장치와 연관된 두 개의 기본 동작, 소위 인출(혹은 읽기)과 저장(혹은 쓰기)이라고 하는 동작들이 필요한 것이다

### 인출(Fetch) 동작 시에는
- #### 주기억장치 특정 번지의 내용을 중앙처리장치 내로 읽어오기 때문에 읽혀진 곳에서는 아무런 변화가 없다
- #### 이러한 인출 동작을 시작하기 위해서, 중앙처리장치는 원하는 위치의 주소를 주기억장치로 보내면서, 읽기 요청을 하여야 한다
- #### 그 때 주기억장치는 그 주소에 저장되어 있는 데이터를 읽어 중앙처리장치로 보내주게 된다

### 저장(Store) 동작 시에는
- #### 하나의 워드 정보를 중앙처리장치로부터 주기억장치 특정 주소를 갖는 기억장소에 덮어쓰기를 하므로 덮어쓴 값이 저장된다
- #### 따라서 이전 내용은 없어지게 된다
- #### 단, 중앙처리장치는 그 위치에 쓰일 데이터와 더불어 원하는 번지의 주소를 주기억장치로 보내야 하며, 읽기와 마찬가지로 쓰기 요청을 하여야 한다

### 이렇게 동작하는데 있어서 값은 어느 시점에서는 주기억장치 또는 다른 시점에서는 중앙처리장치, 즉 중앙처리장치에 있는 기억장치인 레지스터(Register)에 복사되어 있게 된다
- #### 이러한 값의 위치를 나타내기 위해서 주기억장치에 저장된 값을 나타내기 위해 왼쪽 값(L-Value) 그리고 레지스터에 저장된 값을 나타내기 위해 오른쪽 값(R-Value)개념을 사용하는 것이고, 이 개념들을 잘 숙지해야 한다
